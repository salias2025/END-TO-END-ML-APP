```python
# ============================================
# CELL 1: IMPORT LIBRARIES
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_absolute_error, roc_auc_score, accuracy_score

import shap
import warnings
warnings.filterwarnings('ignore')

# Set style for visualizations
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("✅ All libraries imported successfully")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Scikit-learn version: {sklearn.__version__}")
print(f"SHAP version: {shap.__version__}")
```

```python
# ============================================
# CELL 2: GENERATE REALISTIC DATASET (FINAL)
# ============================================

np.random.seed(42)
n_students = 200

print(f"Generating realistic dataset for {n_students} students...")

# ============================================
# Step 1: Generate trimester grades (9-20 scale, mean ~13)
# ============================================

grade_t1 = np.random.normal(13, 2, n_students)
grade_t1 = np.clip(grade_t1, 9, 20)

grade_t2 = grade_t1 + np.random.normal(0.2, 1, n_students)
grade_t2 = np.clip(grade_t2, 9, 20)

grade_t3 = grade_t2 + np.random.normal(0.2, 1, n_students)
grade_t3 = np.clip(grade_t3, 9, 20)

avg_grade_t1_t3 = (grade_t1 + grade_t2 + grade_t3) / 3
avg_grade_t1_t3 = np.round(avg_grade_t1_t3, 1)

grade_trend = grade_t3 - grade_t1
grade_trend = np.round(grade_trend, 1)

# ============================================
# Step 2: Generate BAC skills (0-10 scale)
# ============================================

# Base skill from grades (maps 9→4, 20→10)
base_skill = 4 + (avg_grade_t1_t3 - 9) * (6 / 11)
base_skill = np.clip(base_skill, 4, 10)

grammar_mastery = base_skill + np.random.normal(0, 0.8, n_students)
grammar_mastery = np.clip(grammar_mastery, 4, 10)
grammar_mastery = np.round(grammar_mastery, 1)

essay_writing_score = base_skill - 0.3 + np.random.normal(0, 1, n_students)
essay_writing_score = np.clip(essay_writing_score, 3, 10)
essay_writing_score = np.round(essay_writing_score, 1)

poetry_analysis_score = base_skill + np.random.normal(0, 0.9, n_students)
poetry_analysis_score = np.clip(poetry_analysis_score, 4, 10)
poetry_analysis_score = np.round(poetry_analysis_score, 1)

prose_analysis_score = base_skill + np.random.normal(0, 0.9, n_students)
prose_analysis_score = np.clip(prose_analysis_score, 4, 10)
prose_analysis_score = np.round(prose_analysis_score, 1)

rhetoric_recognition = base_skill - 0.5 + np.random.normal(0, 1.1, n_students)
rhetoric_recognition = np.clip(rhetoric_recognition, 3, 9.5)
rhetoric_recognition = np.round(rhetoric_recognition, 1)

text_comprehension_score = base_skill + 0.3 + np.random.normal(0, 0.8, n_students)
text_comprehension_score = np.clip(text_comprehension_score, 4, 10)
text_comprehension_score = np.round(text_comprehension_score, 1)

# ============================================
# Step 3: Study behavior
# ============================================
essays_per_week = np.random.choice([0,1,2,3,4,5], n_students, p=[0.05,0.1,0.3,0.3,0.15,0.1])
bac_exams_practiced = np.random.poisson(5, n_students)
bac_exams_practiced = np.clip(bac_exams_practiced, 0, 20)
study_hours_per_week = np.random.normal(5, 2, n_students)
study_hours_per_week = np.clip(study_hours_per_week, 2, 10)
study_hours_per_week = np.round(study_hours_per_week, 1)
study_consistency = np.random.choice([1,2,3,4,5], n_students, p=[0.05,0.1,0.2,0.35,0.3])
private_tutoring = np.random.choice([0,1], n_students, p=[0.7,0.3])
class_participation = np.random.choice([1,2,3,4,5], n_students, p=[0.05,0.1,0.25,0.35,0.25])
study_method = np.random.choice([0,1,2], n_students, p=[0.35,0.4,0.25])

# ============================================
# Step 4: Filiere adjustments
# ============================================
filiere = np.random.choice([0,1,2], n_students, p=[0.5,0.3,0.2])

for i in range(n_students):
    if filiere[i] == 0:  # Sciences
        grammar_mastery[i] = min(grammar_mastery[i] + 0.3, 10)
        poetry_analysis_score[i] = max(poetry_analysis_score[i] - 0.3, 4)
        prose_analysis_score[i] = max(prose_analysis_score[i] - 0.3, 4)
    elif filiere[i] == 2:  # Literature
        poetry_analysis_score[i] = min(poetry_analysis_score[i] + 0.5, 10)
        prose_analysis_score[i] = min(prose_analysis_score[i] + 0.5, 10)
        grammar_mastery[i] = max(grammar_mastery[i] - 0.2, 4)

# ============================================
# Step 5: Psychological factors (0-5 scale)
# ============================================
confidence_level = np.random.choice([1,2,3,4,5], n_students, p=[0.05,0.1,0.3,0.35,0.2])
stress_level = np.random.choice([1,2,3,4,5], n_students, p=[0.05,0.1,0.25,0.35,0.25])
interest_in_arabic = np.random.choice([1,2,3,4,5], n_students, p=[0.05,0.1,0.25,0.35,0.25])

# ============================================
# Step 6: Generate BAC score (mean ~13, correlated with grades + consistency)
# ============================================

# Start from trimester average (this ensures mean ~13)
bac_score = avg_grade_t1_t3.copy()

# Add small random variation (±1.5 points max)
bac_score = bac_score + np.random.normal(0, 0.8, n_students)

# Slight adjustment for consistency (consistent students do a bit better)
bac_score = bac_score + (study_consistency - 3) * 0.2

# Slight adjustment for study hours
bac_score = bac_score + (study_hours_per_week - 5) * 0.1

# Slight adjustment for confidence
bac_score = bac_score + (confidence_level - 3) * 0.15

# Clip to 9-20 range (no grades below 9)
bac_score = np.clip(bac_score, 9, 20)

# Round to 1 decimal
bac_score = np.round(bac_score, 1)

# Round grades
grade_t1 = np.round(grade_t1, 1)
grade_t2 = np.round(grade_t2, 1)
grade_t3 = np.round(grade_t3, 1)

# ============================================
# Create DataFrame
# ============================================

df = pd.DataFrame({
    # Academic History
    'grade_t1': grade_t1,
    'grade_t2': grade_t2,
    'grade_t3': grade_t3,
    'avg_grade_t1_t3': avg_grade_t1_t3,
    'grade_trend': grade_trend,
    
    # BAC Exam Skills (0-10 scale)
    'grammar_mastery': grammar_mastery,
    'essay_writing_score': essay_writing_score,
    'poetry_analysis_score': poetry_analysis_score,
    'prose_analysis_score': prose_analysis_score,
    'rhetoric_recognition': rhetoric_recognition,
    'text_comprehension_score': text_comprehension_score,
    
    # Study Behavior
    'essays_per_week': essays_per_week,
    'bac_exams_practiced': bac_exams_practiced,
    'study_hours_per_week': study_hours_per_week,
    'study_consistency': study_consistency,
    'private_tutoring': private_tutoring,
    'class_participation': class_participation,
    'study_method': study_method,
    
    # Context
    'filiere': filiere,
    
    # Psychological (0-5 scale)
    'confidence_level': confidence_level,
    'stress_level': stress_level,
    'interest_in_arabic': interest_in_arabic,
    
    # Target
    'bac_score': bac_score
})

print(f"✅ Dataset generated: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"\n📊 Trimester Grades Statistics:")
print(f"   Grade T1 - Mean: {df['grade_t1'].mean():.1f}, Min: {df['grade_t1'].min():.1f}, Max: {df['grade_t1'].max():.1f}")
print(f"   Grade T2 - Mean: {df['grade_t2'].mean():.1f}, Min: {df['grade_t2'].min():.1f}, Max: {df['grade_t2'].max():.1f}")
print(f"   Grade T3 - Mean: {df['grade_t3'].mean():.1f}, Min: {df['grade_t3'].min():.1f}, Max: {df['grade_t3'].max():.1f}")

print(f"\n📊 BAC Score Statistics:")
print(f"   Mean: {df['bac_score'].mean():.1f}")
print(f"   Min: {df['bac_score'].min():.1f}")
print(f"   Max: {df['bac_score'].max():.1f}")
print(f"   Std: {df['bac_score'].std():.1f}")

print(f"\n📊 BAC Score Distribution (first 30 values):")
print(df['bac_score'].head(30).tolist())

print(f"\n📊 Sample of good scores (≥15):")
good_scores = df[df['bac_score'] >= 15]['bac_score'].tolist()
print(f"   {good_scores[:15]}")

print(f"\n📊 First 5 rows:")
print(df.head())
```

```python
# ============================================
# CELL 3: EXPORT DATASET TO CSV
# ============================================

# Save the dataset to CSV file
csv_filename = 'arabic_bac_dataset.csv'

df.to_csv(csv_filename, index=False)

print(f"✅ Dataset saved successfully as: {csv_filename}")
print(f"📊 File info: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"📁 File location: {csv_filename}")

```

```python
# ============================================
# CELL 6: EDA — BAC SCORE DISTRIBUTION (NORMALITY CHECK)
# ============================================

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Create figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ============================================
# Plot 1: Histogram with KDE (normal curve overlay)
# ============================================
axes[0].hist(df['bac_score'], bins=15, density=True, edgecolor='black', alpha=0.7, color='steelblue', label='Actual Data')

# Generate normal curve with same mean and std
mean = df['bac_score'].mean()
std = df['bac_score'].std()
x = np.linspace(mean - 4*std, mean + 4*std, 100)
normal_curve = stats.norm.pdf(x, mean, std)
axes[0].plot(x, normal_curve, 'r-', linewidth=2, label=f'Normal Curve (μ={mean:.1f}, σ={std:.1f})')

axes[0].set_xlabel('BAC Score (0-20)', fontsize=12)
axes[0].set_ylabel('Density', fontsize=12)
axes[0].set_title('Histogram with Normal Curve Overlay', fontsize=14)
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# ============================================
# Plot 2: Q-Q Plot (Quantile-Quantile)
# ============================================
stats.probplot(df['bac_score'], dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot (Normality Check)', fontsize=14)
axes[1].grid(True, alpha=0.3)

# ============================================
# Plot 3: Boxplot with distribution info
# ============================================
bp = axes[2].boxplot(df['bac_score'], vert=True, patch_artist=True, 
                       boxprops=dict(facecolor='lightblue', color='black'),
                       whiskerprops=dict(color='black'),
                       capprops=dict(color='black'),
                       medianprops=dict(color='red', linewidth=2),
                       flierprops=dict(marker='o', markerfacecolor='red', markersize=6))

axes[2].set_ylabel('BAC Score (0-20)', fontsize=12)
axes[2].set_title('Boxplot with Outliers', fontsize=14)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================
# Normality Tests
# ============================================
print("\n" + "=" * 60)
print("📊 NORMALITY TESTS")
print("=" * 60)

# Shapiro-Wilk test
shapiro_stat, shapiro_p = stats.shapiro(df['bac_score'])
print(f"\nShapiro-Wilk Test:")
print(f"   Statistic: {shapiro_stat:.4f}")
print(f"   P-value:   {shapiro_p:.4f}")
if shapiro_p > 0.05:
    print("   ✅ Data looks normal (p > 0.05)")
else:
    print("   ⚠️ Data may not be normal (p < 0.05)")

# Kolmogorov-Smirnov test
ks_stat, ks_p = stats.kstest(df['bac_score'], 'norm', args=(mean, std))
print(f"\nKolmogorov-Smirnov Test:")
print(f"   Statistic: {ks_stat:.4f}")
print(f"   P-value:   {ks_p:.4f}")
if ks_p > 0.05:
    print("   ✅ Data looks normal (p > 0.05)")
else:
    print("   ⚠️ Data may not be normal (p < 0.05)")

# ============================================
# Distribution Statistics
# ============================================
print("\n" + "=" * 60)
print("📊 DISTRIBUTION STATISTICS")
print("=" * 60)
print(f"Mean:           {mean:.2f}")
print(f"Median:         {df['bac_score'].median():.2f}")
print(f"Mode:           {df['bac_score'].mode().values[0]:.2f}")
print(f"Std Dev:        {std:.2f}")
print(f"Skewness:       {df['bac_score'].skew():.3f}")
print(f"Kurtosis:       {df['bac_score'].kurtosis():.3f}")
print(f"Min:            {df['bac_score'].min():.2f}")
print(f"Max:            {df['bac_score'].max():.2f}")

# Interpretation
skew = df['bac_score'].skew()
kurt = df['bac_score'].kurtosis()

print("\n📊 INTERPRETATION:")
if abs(skew) < 0.5:
    print(f"   Skewness ({skew:.3f}): Approximately symmetric")
elif skew > 0:
    print(f"   Skewness ({skew:.3f}): Right-skewed (tail on right)")
else:
    print(f"   Skewness ({skew:.3f}): Left-skewed (tail on left)")

if abs(kurt) < 1:
    print(f"   Kurtosis ({kurt:.3f}): Similar to normal (mesokurtic)")
elif kurt > 1:
    print(f"   Kurtosis ({kurt:.3f}): Sharper peak than normal (leptokurtic)")
else:
    print(f"   Kurtosis ({kurt:.3f}): Flatter than normal (platykurtic)")
```
