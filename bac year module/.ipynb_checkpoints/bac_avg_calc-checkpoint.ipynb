```python
from IPython.display import display, HTML
import ipywidgets as widgets

# Define the data for each stream
streams_data = {
    "علوم تجريبية": [
        {"subject": "اللغة العربية", "coef": 3, "grade": "___"},
        {"subject": "اللغة الإنجليزية", "coef": 2, "grade": "___"},
        {"subject": "اللغة الفرنسية", "coef": 2, "grade": "___"},
        {"subject": "الرياضيات", "coef": 7, "grade": "___"},
        {"subject": "العلوم الطبيعية", "coef": 6, "grade": "___"},
        {"subject": "العلوم الفيزيائية", "coef": 5, "grade": "___"},
        {"subject": "التاريخ والجغرافيا", "coef": 2, "grade": "___"},
        {"subject": "العلوم الإسلامية", "coef": 2, "grade": "___"},
        {"subject": "الفلسفة", "coef": 2, "grade": "___"},
        {"subject": "التربية البدنية", "coef": 1, "grade": "___"},
        {"subject": "الأمازيغية", "coef": 2, "grade": "___"}
    ],
    "رياضيات": [
        {"subject": "اللغة العربية", "coef": 3, "grade": "___"},
        {"subject": "اللغة الإنجليزية", "coef": 2, "grade": "___"},
        {"subject": "اللغة الفرنسية", "coef": 2, "grade": "___"},
        {"subject": "الرياضيات", "coef": 6, "grade": "___"},
        {"subject": "العلوم الفيزيائية", "coef": 6, "grade": "___"},
        {"subject": "التاريخ والجغرافيا", "coef": 2, "grade": "___"},
        {"subject": "العلوم الإسلامية", "coef": 2, "grade": "___"},
        {"subject": "الفلسفة", "coef": 2, "grade": "___"},
        {"subject": "التربية البدنية", "coef": 1, "grade": "___"},
        {"subject": "الأمازيغية", "coef": 2, "grade": "___"}
    ],
    "تقني رياضي": [
        {"subject": "اللغة العربية", "coef": 3, "grade": "___"},
        {"subject": "اللغة الإنجليزية", "coef": 2, "grade": "___"},
        {"subject": "اللغة الفرنسية", "coef": 2, "grade": "___"},
        {"subject": "الرياضيات", "coef": 5, "grade": "___"},
        {"subject": "العلوم الفيزيائية", "coef": 5, "grade": "___"},
        {"subject": "التكنولوجيا", "coef": 7, "grade": "___"},
        {"subject": "التاريخ والجغرافيا", "coef": 2, "grade": "___"},
        {"subject": "العلوم الإسلامية", "coef": 2, "grade": "___"},
        {"subject": "الفلسفة", "coef": 2, "grade": "___"},
        {"subject": "التربية البدنية", "coef": 1, "grade": "___"},
        {"subject": "الأمازيغية", "coef": 2, "grade": "___"}
    ],
    "تسيير واقتصاد": [
        {"subject": "اللغة العربية", "coef": 3, "grade": "___"},
        {"subject": "اللغة الإنجليزية", "coef": 2, "grade": "___"},
        {"subject": "اللغة الفرنسية", "coef": 2, "grade": "___"},
        {"subject": "الرياضيات", "coef": 2, "grade": "___"},
        {"subject": "التاريخ والجغرافيا", "coef": 4, "grade": "___"},
        {"subject": "العلوم الإسلامية", "coef": 2, "grade": "___"},
        {"subject": "الفلسفة", "coef": 2, "grade": "___"},
        {"subject": "الاقتصاد والمناجمنت", "coef": 5, "grade": "___"},
        {"subject": "القانون", "coef": 2, "grade": "___"},
        {"subject": "التسيير المحاسبي والمالي", "coef": 6, "grade": "___"},
        {"subject": "التربية البدنية", "coef": 1, "grade": "___"},
        {"subject": "الأمازيغية", "coef": 2, "grade": "___"}
    ],
    "لغات أجنبية": [
        {"subject": "اللغة العربية", "coef": 5, "grade": "___"},
        {"subject": "اللغة الإنجليزية", "coef": 5, "grade": "___"},
        {"subject": "اللغة الفرنسية", "coef": 4, "grade": "___"},
        {"subject": "اللغة الألمانية/الإسبانية", "coef": 5, "grade": "___"},
        {"subject": "الرياضيات", "coef": 2, "grade": "___"},
        {"subject": "التاريخ والجغرافيا", "coef": 2, "grade": "___"},
        {"subject": "العلوم الإسلامية", "coef": 2, "grade": "___"},
        {"subject": "الفلسفة", "coef": 2, "grade": "___"},
        {"subject": "التربية البدنية", "coef": 1, "grade": "___"},
        {"subject": "الأمازيغية", "coef": 2, "grade": "___"}
    ],
    "آداب وفلسفة": [
        {"subject": "اللغة العربية", "coef": 7, "grade": "___"},
        {"subject": "اللغة الإنجليزية", "coef": 3, "grade": "___"},
        {"subject": "اللغة الفرنسية", "coef": 3, "grade": "___"},
        {"subject": "الرياضيات", "coef": 2, "grade": "___"},
        {"subject": "التاريخ والجغرافيا", "coef": 4, "grade": "___"},
        {"subject": "العلوم الإسلامية", "coef": 2, "grade": "___"},
        {"subject": "الفلسفة", "coef": 6, "grade": "___"},
        {"subject": "التربية البدنية", "coef": 1, "grade": "___"},
        {"subject": "الأمازيغية", "coef": 2, "grade": "___"}
    ]
}

# Function to generate HTML table for a stream
def generate_table(stream_name, data):
    html = f"""
    <div style="font-family: 'Cairo', 'Tahoma', sans-serif; direction: rtl; margin: 20px auto; max-width: 800px;">
        <div style="background: linear-gradient(135deg, #1e3c72 0%, #2a5298 100%); 
                    color: white; padding: 15px; border-radius: 15px 15px 0 0; text-align: center;">
            <h2 style="margin: 0;">📊 شعبة {stream_name}</h2>
            <p style="margin: 5px 0 0 0; opacity: 0.9;">سجل نقاطك في الجدول أدناه</p>
        </div>
        <div style="overflow-x: auto;">
            <table style="width: 100%; border-collapse: collapse; background: white; box-shadow: 0 4px 8px rgba(0,0,0,0.1);">
                <thead>
                    <tr style="background: #2c3e50; color: white;">
                        <th style="padding: 12px; text-align: center; border: 1px solid #34495e;">📖 المادة</th>
                        <th style="padding: 12px; text-align: center; border: 1px solid #34495e;">⭐ المعامل</th>
                        <th style="padding: 12px; text-align: center; border: 1px solid #34495e;">🎯 النقطة (0-20)</th>
                    </tr>
                </thead>
                <tbody>
    """
    
    for i, item in enumerate(data):
        bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"
        html += f"""
                    <tr style="background: {bg_color};">
                        <td style="padding: 10px; text-align: right; border: 1px solid #dee2e6; font-weight: 500;">{item['subject']}</td>
                        <td style="padding: 10px; text-align: center; border: 1px solid #dee2e6; font-weight: bold; color: #e67e22;">{item['coef']}</td>
                        <td style="padding: 8px; text-align: center; border: 1px solid #dee2e6;">
                            <input type="text" style="width: 80px; padding: 8px; text-align: center; border: 2px solid #3498db; border-radius: 8px; font-size: 14px;" placeholder="__" value="{item['grade'] if item['grade'] != '___' else ''}">
                        </td>
                    </tr>
        """
    
    total_coef = sum(item['coef'] for item in data)
    
    html += f"""
                    <tr style="background: #e8f8f5; font-weight: bold;">
                        <td style="padding: 12px; text-align: right; border: 1px solid #dee2e6;">📅 المجموع</td>
                        <td style="padding: 12px; text-align: center; border: 1px solid #dee2e6; color: #27ae60; font-size: 18px;">{total_coef}</td>
                        <td style="padding: 12px; text-align: center; border: 1px solid #dee2e6;">
                            <button onclick="calculateAverage(this)" style="background: #27ae60; color: white; border: none; padding: 8px 20px; border-radius: 8px; cursor: pointer; font-size: 14px;">
                                📊 احسب المعدل
                            </button>
                        </td>
                    </tr>
                </tbody>
            </table>
        </div>
    </div>
    """
    
    return html

# Create dropdown for stream selection
stream_dropdown = widgets.Dropdown(
    options=list(streams_data.keys()),
    value="تسيير واقتصاد",
    description='🇩🇿 الشعبة:',
    style={'description_width': '80px'},
    layout=widgets.Layout(width='400px', margin='10px auto')
)

# Output area
output = widgets.Output()

# JavaScript function for calculation
js_code = """
<script>
function calculateAverage(button) {
    var row = button.parentElement.parentElement;
    var table = row.parentElement.parentElement.parentElement;
    var rows = table.getElementsByTagName('tr');
    var totalPoints = 0;
    var totalCoef = 0;
    
    for (var i = 0; i < rows.length - 1; i++) {
        var cells = rows[i].getElementsByTagName('td');
        if (cells.length >= 3) {
            var coef = parseFloat(cells[1].innerText);
            var gradeInput = cells[2].getElementsByTagName('input')[0];
            var grade = parseFloat(gradeInput.value);
            
            if (!isNaN(grade) && grade >= 0 && grade <= 20) {
                totalPoints += grade * coef;
                totalCoef += coef;
            }
        }
    }
    
    if (totalCoef > 0) {
        var average = totalPoints / totalCoef;
        var resultRow = document.createElement('div');
        resultRow.style.cssText = 'margin-top: 20px; padding: 20px; border-radius: 10px; text-align: center; font-family: Cairo, Tahoma, sans-serif;';
        
        var color = average >= 16 ? '#2ecc71' : (average >= 14 ? '#27ae60' : (average >= 12 ? '#f39c12' : (average >= 10 ? '#e67e22' : '#e74c3c')));
        resultRow.style.background = 'linear-gradient(135deg, ' + color + ' 0%, ' + color + '80 100%)';
        resultRow.style.color = 'white';
        
        var emoji = average >= 16 ? '🏆' : (average >= 14 ? '🎉' : (average >= 12 ? '👍' : (average >= 10 ? '📈' : '⚠️')));
        var status = average >= 16 ? 'ممتاز' : (average >= 14 ? 'جيد جداً' : (average >= 12 ? 'جيد' : (average >= 10 ? 'مقبول' : 'يحتاج إلى تحسين')));
        
        resultRow.innerHTML = '<div style=\"font-size: 18px;\">معدلك في البكالوريا</div>' +
                              '<div style=\"font-size: 48px; font-weight: bold;\">' + average.toFixed(2) + '<span style=\"font-size: 20px;\">/20</span></div>' +
                              '<div style=\"font-size: 16px;\">' + emoji + ' ' + status + ' ' + emoji + '</div>';
        
        var existingResult = document.getElementById('average-result');
        if (existingResult) existingResult.remove();
        resultRow.id = 'average-result';
        table.parentElement.appendChild(resultRow);
    } else {
        alert('الرجاء إدخال نقاط صحيحة (من 0 إلى 20)');
    }
}
</script>
"""

# Display the UI
def display_table(change=None):
    output.clear_output(wait=True)
    with output:
        display(HTML(js_code))
        display(HTML(generate_table(stream_dropdown.value, streams_data[stream_dropdown.value])))

# Observe dropdown changes
stream_dropdown.observe(display_table, names='value')

# Initial display
print("=" * 60)
print("🎓 حاسبة معدل البكالوريا - الجزائر")
print("=" * 60)
print("اختر شعبتك ثم أدخل نقاطك في الجدول لحساب معدلك")
print("-" * 60)

display(stream_dropdown)
display_table()
display(output)
```

```python
from IPython.display import display, HTML
import pandas as pd
import numpy as np

# ============================================
# COMPLETE RESOURCE DATABASE
# ============================================

RESOURCE_DB = {
    # ========== ARABIC ==========
    'arabic': {
        'name_ar': '📖 اللغة العربية',
        'predicted_score': 12.5,
        'weaknesses': ['تحليل النصوص', 'القواعد النحوية'],
        'youtube': [
            {'name': 'أبو بكر محمد', 'channel': 'Prof Abou Bakr Moubarek', 'url': 'https://youtube.com/@aboubakrmoubarek'},
            {'name': 'أسامة حيقون', 'channel': 'Higoun Oussama', 'url': 'https://youtube.com/@higounoussama'}
        ],
        'books': [
            {'name': 'كتاب الجوهرة', 'author': 'أبو بكر محمد'}
        ],
        'facebook': [
            {'name': 'Prof Twahriya'},
            {'name': 'Prof Belameri Brahim'}
        ]
    },
    
    # ========== MATHEMATICS ==========
    'mathematics': {
        'name_ar': '🧮 الرياضيات',
        'predicted_score': 13.0,
        'weaknesses': ['الدوال', 'الاحتمالات'],
        'youtube': [
            {'name': 'نور الدين', 'channel': 'Prof Nourredine', 'url': 'https://youtube.com/@nourredine'},
            {'name': 'إنفينيتي', 'channel': 'Infinity Maths', 'url': 'https://youtube.com/@infinitymaths'},
            {'name': 'أمين بروف', 'channel': 'Amine Prof Maths', 'url': 'https://youtube.com/@amineprof'},
            {'name': 'ساتورن', 'channel': 'Saturn (احتمالات)', 'url': 'https://youtube.com/@saturn'}
        ],
        'books': [
            {'name': 'سلسلة الفضية', 'author': 'أستاذ نور الدين'}
        ],
        'facebook': [
            {'name': 'Top Maths'}, {'name': 'Math For All'}, {'name': 'Ard Maths'}, {'name': 'Prof Dramnia Fouad'}
        ],
        'apps': [{'name': 'GeoGebra', 'platform': 'Android/iOS'}]
    },
    
    # ========== PHYSICS ==========
    'physics': {
        'name_ar': '⚡ الفيزياء',
        'predicted_score': 11.0,
        'weaknesses': ['الكهرباء', 'الميكانيك'],
        'youtube': [
            {'name': 'عبد اللطيف', 'channel': 'Prof Abdellatif', 'url': 'https://youtube.com/@abdellatif'},
            {'name': 'محمد الأمين زدون', 'channel': 'Zeddoune Mohamed El Amine', 'url': 'https://youtube.com/@zeddoune'},
            {'name': 'سيد أحمد شلال', 'channel': 'Prof Sid Ahmed Chalal', 'url': 'https://youtube.com/@chalal'}
        ],
        'books': [
            {'name': 'كتاب دولفин'}, {'name': 'Cours DZ Edition 2022'}, {'name': 'تأشيرات النجاح'}
        ],
        'facebook': [
            {'name': 'Fares Fergani Physics'}, {'name': 'Prof Chnayet'}, {'name': 'Physics Stif 2024'},
            {'name': 'ملتقى أساتذة العلوم الفيزيائية'}
        ]
    },
    
    # ========== SVT ==========
    'svt': {
        'name_ar': '🔬 العلوم الطبيعية',
        'predicted_score': 14.0,
        'weaknesses': [],
        'youtube': [
            {'name': 'زينة شريف كتفي', 'channel': 'Ketfi Zina Chrif', 'url': 'https://youtube.com/@ketfi'},
            {'name': 'بروف شاوش', 'channel': 'Prof Chaouche', 'url': 'https://youtube.com/@chaouche'},
            {'name': 'مصطفى بد', 'channel': 'Mustapaha Bdd', 'url': 'https://youtube.com/@mustapaha'}
        ],
        'books': [{'name': 'سلسلة الفضية', 'author': 'كتفي شريف'}]
    },
    
    # ========== HISTORY-GEO ==========
    'histoire_geo': {
        'name_ar': '📜 التاريخ والجغرافيا',
        'predicted_score': 13.5,
        'weaknesses': ['تحليل الوثائق'],
        'youtube': [
            {'name': 'أمين سكول', 'channel': 'Amine School', 'url': 'https://youtube.com/@amineschool'},
            {'name': 'هند شواو', 'channel': 'Hind Chaouaou', 'url': 'https://youtube.com/@hindchaouaou'}
        ],
        'books': [
            {'name': 'كتاب برنان', 'author': 'Prof Bourenane'},
            {'name': 'مطوية خرائط', 'type': 'خرائط تاريخ وجغرافيا'}
        ],
        'facebook': [{'name': 'عقبة بن نافع', 'page': 'Okba Ben Nafe3'}]
    },
    
    # ========== ISLAMIA ==========
    'islamia': {
        'name_ar': '🕌 العلوم الإسلامية',
        'predicted_score': 15.0,
        'weaknesses': [],
        'youtube': [
            {'name': 'نوال بوسعدي', 'channel': 'Bous3adi Nawel', 'url': 'https://youtube.com/@bous3adi'},
            {'name': 'بروف صالحي', 'channel': 'Prof Salhi', 'url': 'https://youtube.com/@salhi'},
            {'name': 'بروف شمس الدين', 'channel': 'Prof Chams Ddine', 'url': 'https://youtube.com/@chams'}
        ],
        'books': [{'name': 'سلسلة الخضراء'}, {'name': 'سلسلة العرجوانية'}],
        'facebook': [{'name': 'tarbia.dz', 'page': 'Tarbia.dz online website'}]
    },
    
    # ========== PHILOSOPHY ==========
    'philosophy': {
        'name_ar': '💭 الفلسفة',
        'predicted_score': 12.0,
        'weaknesses': ['المقال الجدلي'],
        'youtube': [{'name': 'بروف ضيف', 'channel': 'Prof Dhif', 'url': 'https://youtube.com/@dhif'}],
        'books': [{'name': 'كتاب هبة جبابرية', 'author': 'Hiba Djebabria'}]
    },
    
    # ========== DROIT ==========
    'droit': {
        'name_ar': '⚖️ القانون',
        'predicted_score': 10.5,
        'weaknesses': ['قانون الشركات', 'الاستدلال القانوني'],
        'youtube': [
            {'name': 'قناة القانون', 'channel': 'Droit Facile'},
            {'name': 'شرح قانون الشركات', 'channel': 'BAC Droit'}
        ],
        'books': [{'name': 'ملخص القانون', 'author': 'أ. بن عمر'}],
        'facebook': [{'name': 'Prof Droit'}, {'name': 'Groupe BAC Droit'}]
    },
    
    # ========== ECONOMIE ==========
    'economie': {
        'name_ar': '📈 الاقتصاد',
        'predicted_score': 12.5,
        'weaknesses': ['آليات السوق'],
        'youtube': [
            {'name': 'اقتصاد باك', 'channel': 'Economie Bac'},
            {'name': 'الاقتصاد ببساطة', 'channel': 'Eco Simple'}
        ],
        'books': [{'name': 'ملخص الاقتصاد', 'author': 'أ. بوعزة'}],
        'facebook': [{'name': 'Eco BAC'}, {'name': 'Groupe Economie'}]
    },
    
    # ========== GESTION ==========
    'gestion': {
        'name_ar': '📊 التسيير المحاسبي والمالي',
        'predicted_score': 9.5,
        'weaknesses': ['تسيير التكاليف', 'جداول التوزيع', 'الاهتلاكات'],
        'youtube': [
            {'name': 'تسيير المحاسبة', 'channel': 'Compta Facile'},
            {'name': 'تمارين التسيير', 'channel': 'Gestion BAC'}
        ],
        'books': [{'name': 'دليل التسيير', 'author': 'أ. بوجمعة'}],
        'facebook': [{'name': 'Gestion BAC 2025'}, {'name': 'Compta DZ'}]
    },
    
    # ========== ENGLISH ==========
    'english': {
        'name_ar': '🇬🇧 اللغة الإنجليزية',
        'predicted_score': 13.0,
        'weaknesses': ['القواعد', 'الترجمة'],
        'youtube': [
            {'name': 'English Bac', 'channel': 'Bac English'},
            {'name': 'Learn English', 'channel': 'English DZ'}
        ],
        'books': [{'name': 'English Summary', 'author': 'Prof English'}],
        'apps': [{'name': 'Duolingo'}, {'name': 'Memrise'}]
    },
    
    # ========== FRENCH ==========
    'french': {
        'name_ar': '🇫🇷 اللغة الفرنسية',
        'predicted_score': 12.5,
        'weaknesses': ['المقال', 'القواعد'],
        'youtube': [
            {'name': 'Français Bac', 'channel': 'Bac Francais'},
            {'name': 'Méthodologie', 'channel': 'French DZ'}
        ],
        'books': [{'name': 'Résumé Français', 'author': 'Prof Français'}]
    },
    
    # ========== TAMAZIGHT ==========
    'tamazight': {
        'name_ar': 'ⵜⴰⵎⴰⵣⵉⵖⵜ الأمازيغية',
        'predicted_score': 14.0,
        'weaknesses': [],
        'youtube': [{'name': 'Tamazight Bac', 'channel': 'Tifinagh Education'}],
        'facebook': [{'name': 'Tamazight BAC'}]
    },
    
    # ========== TECHNOLOGY ==========
    'technology': {
        'name_ar': '💻 التكنولوجيا',
        'predicted_score': 11.5,
        'weaknesses': ['الرسم التقني', 'الميكانيك'],
        'youtube': [{'name': 'Techno Bac', 'channel': 'Technologie DZ'}],
        'facebook': [{'name': 'Groupe Techno BAC'}]
    },
    
    # ========== GENERAL RESOURCES ==========
    'general': {
        'facebook': [
            {'name': 'Prof Bouznaq Remdan'}, {'name': 'Prof Bouchnaq Youcef'},
            {'name': 'Prof Khaled Bakhchcha'}, {'name': 'Groupes BAC 2025'},
            {'name': 'PDF trouve sur facebook'}
        ],
        'instagram': [{'name': 'Reels - astuces, erreurs communes, quizzes', 'tip': 'تابع الـ reels'}, {'name': 'Bac DZ Officiel'}],
        'apps': [{'name': 'تطبيق BAC sujets + solutions'}, {'name': 'تطبيق Quiz'}, {'name': 'Telegram'}],
        'courses': [{'name': 'Cours particulier (presentiel)'}, {'name': 'Dawarat présentiel'}, {'name': 'Dawarat sur Zoom'}]
    }
}

# ============================================
# HELPER FUNCTIONS
# ============================================

def get_priority(predicted_score, weakness_count):
    """Determine priority level based on predicted score and weaknesses"""
    if predicted_score < 10 or weakness_count >= 2:
        return "critical", "#e74c3c", "🔴 حرجة"
    elif predicted_score < 12 or weakness_count >= 1:
        return "high", "#e67e22", "🟠 عالية"
    elif predicted_score < 14:
        return "medium", "#f39c12", "🟡 متوسطة"
    else:
        return "low", "#2ecc71", "🟢 منخفضة"

def get_resources_by_priority(subject_data, priority_key):
    """Return resources based on priority level"""
    if priority_key == "critical":
        # Show ALL resources
        return subject_data
    elif priority_key == "high":
        # Show YouTube + Books only
        return {
            'youtube': subject_data.get('youtube', []),
            'books': subject_data.get('books', [])
        }
    elif priority_key == "medium":
        # Show YouTube only
        return {
            'youtube': subject_data.get('youtube', [])
        }
    else:
        return {}

# ============================================
# GENERATE RECOMMENDATIONS
# ============================================

def generate_recommendations():
    """Generate all recommendations based on hardcoded data"""
    
    all_subjects = []
    
    for subject_id, data in RESOURCE_DB.items():
        if subject_id == 'general':
            continue
        
        weakness_count = len(data.get('weaknesses', []))
        predicted_score = data.get('predicted_score', 12)
        priority_key, priority_color, priority_text = get_priority(predicted_score, weakness_count)
        
        resources = get_resources_by_priority(data, priority_key)
        
        all_subjects.append({
            'id': subject_id,
            'name': data['name_ar'],
            'score': predicted_score,
            'weakness_count': weakness_count,
            'weaknesses': data.get('weaknesses', []),
            'priority_text': priority_text,
            'priority_key': priority_key,
            'priority_color': priority_color,
            'resources': resources
        })
    
    # Sort by priority (critical first, then high, then medium, then low)
    priority_order = {'critical': 0, 'high': 1, 'medium': 2, 'low': 3}
    all_subjects.sort(key=lambda x: priority_order[x['priority_key']])
    
    return all_subjects, RESOURCE_DB.get('general', {})

# ============================================
# HTML/CSS DISPLAY
# ============================================

def display_recommendations():
    """Display beautiful HTML recommendations"""
    
    recommendations, general_resources = generate_recommendations()
    
    html = f"""
    <!DOCTYPE html>
    <html dir="rtl" lang="ar">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>منصة توصيات الموارد التعليمية - البكالوريا</title>
        <style>
            @import url('https://fonts.googleapis.com/css2?family=Cairo:wght@400;500;600;700;800&display=swap');
            
            * {{
                margin: 0;
                padding: 0;
                box-sizing: border-box;
            }}
            
            body {{
                font-family: 'Cairo', 'Tahoma', sans-serif;
                background: linear-gradient(135deg, #f5f7fa 0%, #e9ecef 100%);
                padding: 30px;
            }}
            
            .container {{
                max-width: 1400px;
                margin: 0 auto;
            }}
            
            /* Header */
            .header {{
                background: linear-gradient(135deg, #1e3c72 0%, #2a5298 100%);
                color: white;
                padding: 40px 30px;
                border-radius: 20px;
                text-align: center;
                margin-bottom: 30px;
                box-shadow: 0 10px 30px rgba(0,0,0,0.1);
            }}
            
            .header h1 {{
                font-size: 32px;
                margin-bottom: 10px;
            }}
            
            .header p {{
                font-size: 16px;
                opacity: 0.9;
            }}
            
            .badge {{
                background: rgba(255,255,255,0.2);
                border-radius: 30px;
                padding: 8px 20px;
                display: inline-block;
                margin-top: 15px;
                font-size: 14px;
            }}
            
            /* Summary Cards */
            .summary-cards {{
                display: grid;
                grid-template-columns: repeat(auto-fit, minmax(280px, 1fr));
                gap: 20px;
                margin-bottom: 30px;
            }}
            
            .summary-card {{
                background: white;
                border-radius: 20px;
                padding: 20px;
                text-align: center;
                box-shadow: 0 4px 15px rgba(0,0,0,0.08);
                transition: transform 0.3s;
            }}
            
            .summary-card:hover {{
                transform: translateY(-5px);
            }}
            
            .summary-card h3 {{
                font-size: 14px;
                color: #666;
                margin-bottom: 10px;
            }}
            
            .summary-card .number {{
                font-size: 48px;
                font-weight: bold;
                color: #2c3e50;
            }}
            
            /* Priority Table */
            .priority-table {{
                background: white;
                border-radius: 20px;
                overflow: hidden;
                margin-bottom: 30px;
                box-shadow: 0 4px 15px rgba(0,0,0,0.08);
            }}
            
            .priority-table table {{
                width: 100%;
                border-collapse: collapse;
            }}
            
            .priority-table th {{
                background: #2c3e50;
                color: white;
                padding: 15px;
                text-align: center;
            }}
            
            .priority-table td {{
                padding: 12px;
                text-align: center;
                border-bottom: 1px solid #eee;
            }}
            
            .priority-table tr:hover {{
                background: #f8f9fa;
            }}
            
            .critical-badge {{
                background: #e74c3c;
                color: white;
                padding: 4px 12px;
                border-radius: 20px;
                font-size: 12px;
                display: inline-block;
            }}
            
            .high-badge {{
                background: #e67e22;
                color: white;
                padding: 4px 12px;
                border-radius: 20px;
                font-size: 12px;
                display: inline-block;
            }}
            
            .medium-badge {{
                background: #f39c12;
                color: white;
                padding: 4px 12px;
                border-radius: 20px;
                font-size: 12px;
                display: inline-block;
            }}
            
            .low-badge {{
                background: #2ecc71;
                color: white;
                padding: 4px 12px;
                border-radius: 20px;
                font-size: 12px;
                display: inline-block;
            }}
            
            /* Resource Cards */
            .resource-section {{
                margin-bottom: 40px;
            }}
            
            .section-title {{
                font-size: 24px;
                font-weight: 700;
                color: #2c3e50;
                margin-bottom: 20px;
                padding-right: 15px;
                border-right: 5px solid #e74c3c;
            }}
            
            .resource-grid {{
                display: grid;
                grid-template-columns: repeat(auto-fill, minmax(350px, 1fr));
                gap: 25px;
            }}
            
            .resource-card {{
                background: white;
                border-radius: 20px;
                overflow: hidden;
                box-shadow: 0 10px 25px rgba(0,0,0,0.08);
                transition: all 0.3s;
                cursor: pointer;
                border: 1px solid rgba(0,0,0,0.05);
            }}
            
            .resource-card:hover {{
                transform: translateY(-5px);
                box-shadow: 0 20px 35px rgba(0,0,0,0.15);
            }}
            
            .card-header {{
                padding: 20px;
                color: white;
                position: relative;
            }}
            
            .card-icon {{
                font-size: 40px;
                margin-bottom: 10px;
            }}
            
            .card-title {{
                font-size: 18px;
                font-weight: 700;
                margin-bottom: 5px;
            }}
            
            .card-subtitle {{
                font-size: 12px;
                opacity: 0.9;
            }}
            
            .card-body {{
                padding: 20px;
            }}
            
            .card-description {{
                color: #666;
                font-size: 14px;
                line-height: 1.6;
                margin-bottom: 15px;
            }}
            
            .card-meta {{
                display: flex;
                justify-content: space-between;
                align-items: center;
                border-top: 1px solid #eee;
                padding-top: 15px;
                margin-top: 10px;
            }}
            
            .tag {{
                background: #ecf0f1;
                padding: 4px 12px;
                border-radius: 20px;
                font-size: 11px;
                color: #2c3e50;
            }}
            
            .card-footer {{
                padding: 15px 20px;
                background: #f8f9fa;
                border-top: 1px solid #eee;
                text-align: center;
            }}
            
            .btn-link {{
                background: #3498db;
                color: white;
                border: none;
                padding: 10px 25px;
                border-radius: 30px;
                cursor: pointer;
                font-family: 'Cairo', sans-serif;
                font-size: 14px;
                transition: all 0.3s;
                text-decoration: none;
                display: inline-block;
            }}
            
            .btn-link:hover {{
                background: #2980b9;
                transform: scale(1.05);
            }}
            
            .weakness-tag {{
                background: #ffeaa7;
                color: #d63031;
                padding: 4px 12px;
                border-radius: 20px;
                font-size: 11px;
                display: inline-block;
                margin-left: 8px;
                margin-bottom: 8px;
            }}
            
            .footer {{
                text-align: center;
                padding: 30px;
                color: #888;
                font-size: 12px;
            }}
            
            @media (max-width: 768px) {{
                .resource-grid {{
                    grid-template-columns: 1fr;
                }}
            }}
        </style>
    </head>
    <body>
    <div class="container">
        <!-- Header -->
        <div class="header">
            <h1>📚 منصة توصيات الموارد التعليمية</h1>
            <p>أفضل القنوات، الكتب، التطبيقات، والدورات لمساعدتك في البكالوريا</p>
            <div class="badge">🎯 توصيات مبنية على تحليل أداء الطلاب</div>
        </div>
        
        <!-- Summary Cards -->
        <div class="summary-cards">
            <div class="summary-card">
                <h3>📚 عدد المواد</h3>
                <div class="number">{len(recommendations)}</div>
            </div>
            <div class="summary-card">
                <h3>🔴 مواد تحتاج تركيزاً عالياً</h3>
                <div class="number">{sum(1 for r in recommendations if r['priority_key'] == 'critical')}</div>
            </div>
            <div class="summary-card">
                <h3>📖 إجمالي الموارد الموصى بها</h3>
                <div class="number">{sum(len(r['resources'].get('youtube', [])) + len(r['resources'].get('books', [])) + len(r['resources'].get('facebook', [])) + len(r['resources'].get('apps', [])) for r in recommendations)}</div>
            </div>
        </div>
        
        <!-- Priority Table -->
        <div class="priority-table">
            <table>
                <thead>
                    <tr><th>#</th><th>المادة</th><th>النقطة المتوقعة</th><th>نقاط الضعف</th><th>أولوية التركيز</th></tr>
                </thead>
                <tbody>
    """
    
    for i, r in enumerate(recommendations, 1):
        badge_class = r['priority_key']
        weaknesses_display = ', '.join(r['weaknesses']) if r['weaknesses'] else '—'
        
        html += f"""
                    <tr>
                        <td>{i}</td>
                        <td style="text-align: right; font-weight: 600;">{r['name']}</td>
                        <td>{r['score']}/20</td>
                        <td style="text-align: right;">{weaknesses_display}</td>
                        <td><span class="{badge_class}-badge">{r['priority_text']}</span></td>
                    </tr>
        """
    
    html += """
                </tbody>
            </table>
        </div>
        
        <!-- Resources by Subject -->
    """
    
    for r in recommendations:
        if not r['resources']:
            continue
            
        priority_icon = "🔴" if r['priority_key'] == "critical" else "🟠" if r['priority_key'] == "high" else "🟡" if r['priority_key'] == "medium" else "🟢"
        
        html += f"""
        <div class="resource-section">
            <div class="section-title">
                {priority_icon} {r['name']} - {r['priority_text']}
                <span style="font-size: 14px; font-weight: normal; color: #666; margin-right: 10px;">(النقطة المتوقعة: {r['score']}/20)</span>
            </div>
        """
        
        if r['weaknesses']:
            html += f"""
            <div style="margin-bottom: 15px;">
                <strong>⚠️ نقاط الضعف:</strong><br>
            """
            for w in r['weaknesses']:
                html += f'<span class="weakness-tag">{w}</span>'
            html += "</div>"
        
        html += '<div class="resource-grid">'
        
        # YouTube Resources
        if r['resources'].get('youtube'):
            for yt in r['resources']['youtube']:
                header_color = "#e74c3c"
                html += f"""
                <div class="resource-card">
                    <div class="card-header" style="background: linear-gradient(135deg, {header_color} 0%, {header_color}cc 100%);">
                        <div class="card-icon">🎬</div>
                        <div class="card-title">{yt['name']}</div>
                        <div class="card-subtitle">{yt.get('channel', 'قناة يوتيوب')}</div>
                    </div>
                    <div class="card-body">
                        <div class="card-description">قناة متخصصة في شرح {r['name']} مع تمارين تطبيقية.</div>
                        <div class="card-meta">
                            <div class="tag">📺 يوتيوب</div>
                            <div class="tag">⭐ موصى به</div>
                        </div>
                    </div>
                    <div class="card-footer">
                        <button class="btn-link" onclick="window.open('{yt.get('url', '#')}', '_blank')">🔗 زيارة القناة</button>
                    </div>
                </div>
                """
        
        # Books
        if r['resources'].get('books'):
            for book in r['resources']['books']:
                header_color = "#27ae60"
                html += f"""
                <div class="resource-card">
                    <div class="card-header" style="background: linear-gradient(135deg, {header_color} 0%, {header_color}cc 100%);">
                        <div class="card-icon">📖</div>
                        <div class="card-title">{book['name']}</div>
                        <div class="card-subtitle">{book.get('author', 'كتاب تعليمي')}</div>
                    </div>
                    <div class="card-body">
                        <div class="card-description">كتاب ملخص وشامل لمادة {r['name']} مع تمارين محلولة.</div>
                        <div class="card-meta">
                            <div class="tag">📚 كتاب</div>
                            <div class="tag">⭐ ملخص</div>
                        </div>
                    </div>
                    <div class="card-footer">
                        <button class="btn-link" onclick="alert('📖 يمكنك الحصول على هذا الكتاب من المكتبات أو PDF على فيسبوك')">🔗 معلومات عن الكتاب</button>
                    </div>
                </div>
                """
        
        # Facebook Pages
        if r['resources'].get('facebook'):
            for fb in r['resources']['facebook']:
                header_color = "#3b5998"
                html += f"""
                <div class="resource-card">
                    <div class="card-header" style="background: linear-gradient(135deg, {header_color} 0%, {header_color}cc 100%);">
                        <div class="card-icon">🌐</div>
                        <div class="card-title">{fb['name']}</div>
                        <div class="card-subtitle">صفحة فيسبوك</div>
                    </div>
                    <div class="card-body">
                        <div class="card-description">صفحة متخصصة في تقديم ملخصات وتمارين وامتحانات لمادة {r['name']}.</div>
                        <div class="card-meta">
                            <div class="tag">📘 فيسبوك</div>
                            <div class="tag">⭐ متابعة يومية</div>
                        </div>
                    </div>
                    <div class="card-footer">
                        <button class="btn-link" onclick="alert('🔍 ابحث عن {fb['name']} على فيسبوك')">🔗 زيارة الصفحة</button>
                    </div>
                </div>
                """
        
        # Apps
        if r['resources'].get('apps'):
            for app in r['resources']['apps']:
                header_color = "#3498db"
                html += f"""
                <div class="resource-card">
                    <div class="card-header" style="background: linear-gradient(135deg, {header_color} 0%, {header_color}cc 100%);">
                        <div class="card-icon">📱</div>
                        <div class="card-title">{app['name']}</div>
                        <div class="card-subtitle">{app.get('platform', 'تطبيق')}</div>
                    </div>
                    <div class="card-body">
                        <div class="card-description">تطبيق مفيد لمراجعة مادة {r['name']} واختبار مستواك.</div>
                        <div class="card-meta">
                            <div class="tag">📱 تطبيق</div>
                            <div class="tag">⭐ مجاني</div>
                        </div>
                    </div>
                    <div class="card-footer">
                        <button class="btn-link" onclick="alert('📱 يمكنك تحميل {app['name']} من متجر التطبيقات')">🔗 معلومات التطبيق</button>
                    </div>
                </div>
                """
        
        html += """
            </div>
        </div>
        """
    
    # General Resources
    html += f"""
        <div class="resource-section">
            <div class="section-title">📱 موارد عامة (لجميع المواد)</div>
            <div class="resource-grid">
    """
    
    for fb in general_resources.get('facebook', [])[:4]:
        html += f"""
        <div class="resource-card">
            <div class="card-header" style="background: linear-gradient(135deg, #3b5998 0%, #3b5998cc 100%);">
                <div class="card-icon">🌐</div>
                <div class="card-title">{fb['name']}</div>
                <div class="card-subtitle">صفحة فيسبوك</div>
            </div>
            <div class="card-footer">
                <button class="btn-link" onclick="alert('🔍 ابحث عن {fb['name']} على فيسبوك')">🔗 زيارة الصفحة</button>
            </div>
        </div>
        """
    
    for app in general_resources.get('apps', []):
        html += f"""
        <div class="resource-card">
            <div class="card-header" style="background: linear-gradient(135deg, #3498db 0%, #3498dbcc 100%);">
                <div class="card-icon">📱</div>
                <div class="card-title">{app['name']}</div>
                <div class="card-subtitle">تطبيق</div>
            </div>
            <div class="card-footer">
                <button class="btn-link" onclick="alert('📱 تطبيق {app['name']} متاح على متجر التطبيقات')">🔗 معلومات</button>
            </div>
        </div>
        """
    
    for course in general_resources.get('courses', []):
        html += f"""
        <div class="resource-card">
            <div class="card-header" style="background: linear-gradient(135deg, #9b59b6 0%, #9b59b6cc 100%);">
                <div class="card-icon">🎓</div>
                <div class="card-title">{course['name']}</div>
                <div class="card-subtitle">دورة تدريبية</div>
            </div>
            <div class="card-footer">
                <button class="btn-link" onclick="alert('🎓 دورة {course['name']} - يمكنك التسجيل عبر التواصل مع المنظمين')">🔗 معلومات التسجيل</button>
            </div>
        </div>
        """
    
    html += """
            </div>
        </div>
        
        <div class="footer">
            <p>✨ هذه التوصيات مبنية على تحليل أداء طلاب البكالوريا في السنوات السابقة ✨</p>
            <p>📌 ملاحظة: المواد ذات الأولوية "حرجة" تحتاج إلى تركيز يومي، المواد "عالية" تحتاج إلى متابعة أسبوعية</p>
        </div>
    </div>
    </body>
    </html>
    """
    
    display(HTML(html))

# ============================================
# RUN THE RECOMMENDER
# ============================================

print("=" * 70)
print("📚 منصة توصيات الموارد التعليمية - البكالوريا الجزائرية")
print("=" * 70)
print("✅ تم تحميل جميع المواد بنجاح")
print("✅ تم تحليل نقاط الضعف لكل مادة")
print("✅ تم إنشاء توصيات مخصصة")
print("=" * 70)
print()

display_recommendations()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import math

# ============================================
# DATA
# ============================================

categories = [
    {"name": "🎓 الدروس الخصوصية", "percentage": 40, "color": "red"},
    {"name": "📚 الكتب والوثائق", "percentage": 25, "color": "blue"},
    {"name": "💻 دورات Zoom/Teams", "percentage": 15, "color": "orange"},
    {"name": "📱 أجهزة إلكترونية", "percentage": 8, "color": "purple"},
    {"name": "🚗 المواصلات", "percentage": 5, "color": "green"},
    {"name": "✏️ القرطاسية", "percentage": 4, "color": "pink"},
    {"name": "🖨️ الطباعة", "percentage": 2, "color": "brown"},
    {"name": "🍎 وجبات ومشروبات", "percentage": 1, "color": "gold"}
]

# ============================================
# CREATE WIDGETS
# ============================================

total_budget = widgets.IntText(
    value=500000,
    description='الميزانية الإجمالية (دج):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

# Create sliders and labels for each category
sliders = []
percent_labels = []
amount_labels = []

for i, cat in enumerate(categories):
    slider = widgets.FloatSlider(
        value=cat['percentage'],
        min=0,
        max=100,
        step=1,
        description=cat['name'],
        style={'description_width': '200px'},
        layout=widgets.Layout(width='500px')
    )
    percent_label = widgets.Label(f"{cat['percentage']}%")
    amount_label = widgets.Label("0 دج")
    
    sliders.append(slider)
    percent_labels.append(percent_label)
    amount_labels.append(amount_label)

# Output area
output = widgets.Output()

# ============================================
# UPDATE FUNCTION
# ============================================

def update_all(change=None):
    total = total_budget.value
    
    # Get current percentages from sliders
    percentages = [s.value for s in sliders]
    
    # Normalize to 100%
    total_pct = sum(percentages)
    if total_pct != 100:
        # Adjust the last category
        percentages[-1] = max(0, 100 - sum(percentages[:-1]))
        sliders[-1].value = percentages[-1]
    
    # Update labels and calculate amounts
    total_amount = 0
    for i, pct in enumerate(percentages):
        percent_labels[i].value = f"{pct:.0f}%"
        amount = round((pct / 100) * total)
        amount_labels[i].value = f"{amount:,} دج"
        total_amount += amount
    
    remaining = total - total_amount
    
    with output:
        clear_output(wait=True)
        
        print("=" * 50)
        print(f"💰 الميزانية الإجمالية: {total:,} دج")
        print(f"📊 إجمالي المصروفات: {total_amount:,} دج")
        print(f"💵 المتبقي: {remaining:,} دج")
        print("=" * 50)
        print()
        print("📋 تفاصيل المصروفات:")
        print("-" * 40)
        
        for i, cat in enumerate(categories):
            pct = percentages[i]
            amount = round((pct / 100) * total)
            print(f"{cat['name']:<30} {pct:>3.0f}%   {amount:>12,} دج")
        
        print("-" * 40)
        
        if remaining < 0:
            print("⚠️ تحذير: أنت تتجاوز ميزانيتك!")
        elif remaining > 0:
            print(f"✅ فائض: {remaining:,} دج - يمكنك استثماره في مواد تحتاج تقوية")
        else:
            print("✅ ميزانيتك متوازنة تماماً")

# ============================================
# ATTACH EVENT HANDLERS
# ============================================

total_budget.observe(update_all, names='value')
for slider in sliders:
    slider.observe(update_all, names='value')

# ============================================
# RESET FUNCTION
# ============================================

def reset_percentages(b):
    for i, cat in enumerate(categories):
        sliders[i].value = cat['percentage']
    total_budget.value = 500000

reset_button = widgets.Button(
    description='🔄 إعادة تعيين النسب',
    button_style='warning',
    layout=widgets.Layout(width='250px')
)
reset_button.on_click(reset_percentages)

# ============================================
# SAVE FUNCTION
# ============================================

def save_budget(b):
    total = total_budget.value
    percentages = [s.value for s in sliders]
    
    message = "📊 توزيع الميزانية المحفوظ:\n"
    message += "=" * 40 + "\n"
    
    for i, cat in enumerate(categories):
        amount = round((percentages[i] / 100) * total)
        message += f"{cat['name']}: {percentages[i]:.0f}% ({amount:,} دج)\n"
    
    message += "=" * 40 + "\n"
    message += f"💰 المجموع: {sum([round((s.value/100)*total) for s in sliders]):,} دج\n"
    
    print(message)
    # Also show as alert in HTML context
    display(widgets.HTML(f"<pre style='background:#e8f8f5; padding:15px; border-radius:10px;'>{message}</pre>"))

save_button = widgets.Button(
    description='💾 حفظ التوزيع',
    button_style='success',
    layout=widgets.Layout(width='250px')
)
save_button.on_click(save_budget)

# ============================================
# EXPORT FUNCTION
# ============================================

def export_budget(b):
    total = total_budget.value
    percentages = [s.value for s in sliders]
    
    report = "📊 تقرير ميزانية البكالوريا\n"
    report += "=" * 50 + "\n"
    report += f"📅 التاريخ: {__import__('datetime').datetime.now().strftime('%d/%m/%Y')}\n"
    report += f"💰 الميزانية الإجمالية: {total:,} دج\n"
    report += "=" * 50 + "\n"
    report += "📋 تفاصيل المصروفات:\n"
    
    total_amount = 0
    for i, cat in enumerate(categories):
        amount = round((percentages[i] / 100) * total)
        total_amount += amount
        report += f"   • {cat['name']}: {percentages[i]:.0f}% ({amount:,} دج)\n"
    
    report += "=" * 50 + "\n"
    report += f"💸 إجمالي المصروفات: {total_amount:,} دج\n"
    report += f"💵 المتبقي: {(total - total_amount):,} دج\n"
    
    display(widgets.HTML(f"<pre style='background:#f0f0f0; padding:15px; border-radius:10px;'>{report}</pre>"))

export_button = widgets.Button(
    description='📥 تصدير التقرير',
    button_style='info',
    layout=widgets.Layout(width='250px')
)
export_button.on_click(export_budget)

# ============================================
# DISPLAY EVERYTHING
# ============================================

print("=" * 60)
print("💰 حاسبة ميزانية البكالوريا")
print("=" * 60)
print("يمكنك تعديل النسب المئوية لكل فئة باستخدام شريط التمرير")
print("-" * 60)

display(total_budget)

# Display sliders in a VBox
sliders_box = widgets.VBox([
    widgets.HBox([sliders[i], percent_labels[i], amount_labels[i]]) 
    for i in range(len(categories))
])
display(sliders_box)

# Buttons
buttons_box = widgets.HBox([reset_button, save_button, export_button], layout=widgets.Layout(justify_content='center'))
display(buttons_box)

display(output)

# Initial update
update_all()
```

```python
# CELL 4: ML To-Do List Generator (WITH TASK TYPE COLUMN)
# ============================================
# Run this after Cell 3
# 
# NEW APPROACH:
# ============================================
# Instead of generating ALL task types for every subject,
# we randomly select 1-2 task types per subject per week.
# This creates a compact, realistic to-do list (15-20 tasks).
# ============================================

# ============================================
# 1. HELPER FUNCTIONS
# ============================================

def calculate_priority(subject_data):
    """
    Priority = 90% coefficient + 10% ML score
    Returns: 1-5 stars
    """
    coefficient = subject_data.get('coefficient', 2)
    ml_score = subject_data.get('ml_score', 12)
    is_weak = subject_data.get('weak', False)
    
    # Base from coefficient (90% weight)
    if coefficient >= 7:
        coef_score = 5
    elif coefficient == 6:
        coef_score = 5
    elif coefficient == 5:
        coef_score = 4
    elif coefficient == 4:
        coef_score = 3
    elif coefficient == 3:
        coef_score = 2
    elif coefficient == 2:
        coef_score = 2
    else:
        coef_score = 1
    
    # ML score adjustment (10% weight)
    if ml_score < 10:
        ml_adj = 1.5
    elif ml_score < 11:
        ml_adj = 1.0
    elif ml_score < 12:
        ml_adj = 0.5
    elif ml_score < 13:
        ml_adj = 0
    elif ml_score < 14:
        ml_adj = -0.5
    elif ml_score < 15:
        ml_adj = -1
    else:
        ml_adj = -1.5
    
    # Weak subject boost
    if is_weak:
        ml_adj = ml_adj + 0.5
    
    # Combine
    final = (coef_score * 0.9) + (ml_adj * 0.1)
    
    # Round and cap
    final = int(round(final))
    if final > 5:
        final = 5
    if final < 1:
        final = 1
    
    return final


def get_past_bac_duration(coefficient):
    """
    RULE-BASED: Fixed duration by coefficient
    Past BAC is the ONLY task type with fixed times
    """
    duration_map = {
        7: 4.5,
        6: 4.0,
        5: 3.5,
        4: 3.0,
        3: 2.5,
        2: 2.0,
        1: 1.5
    }
    return duration_map.get(coefficient, 3.0)


def calculate_task_duration(subject_data, task_type, week_focus_multiplier=1.0):
    """
    Calculate duration for a task based on:
    - Coefficient (primary)
    - ML score (adjuster)
    - Task type
    - Week focus multiplier
    """
    coefficient = subject_data.get('coefficient', 2)
    ml_score = subject_data.get('ml_score', 12)
    is_weak = subject_data.get('weak', False)
    
    # Past BAC is rule-based, not dynamic
    if task_type == 'past_bac':
        return get_past_bac_duration(coefficient) * week_focus_multiplier
    
    # Base duration by task type
    if task_type == 'review':
        base = 1.0
    elif task_type == 'exercises':
        base = 1.5
    elif task_type == 'weakness_focus':
        base = 1.0
    else:
        base = 1.0
    
    # Coefficient multiplier
    if coefficient >= 6:
        coef_mult = 1.4
    elif coefficient >= 5:
        coef_mult = 1.2
    elif coefficient >= 4:
        coef_mult = 1.1
    else:
        coef_mult = 1.0
    
    # ML score multiplier (lower score = more time)
    if ml_score < 10:
        ml_mult = 1.5
    elif ml_score < 11:
        ml_mult = 1.4
    elif ml_score < 12:
        ml_mult = 1.2
    elif ml_score < 13:
        ml_mult = 1.0
    elif ml_score < 14:
        ml_mult = 0.9
    elif ml_score < 15:
        ml_mult = 0.8
    else:
        ml_mult = 0.7
    
    # Weak subject boost
    if is_weak and task_type in ['weakness_focus', 'exercises']:
        ml_mult = ml_mult + 0.1
    
    # Calculate
    duration = base * coef_mult * ml_mult * week_focus_multiplier
    
    # Round to nearest 0.5
    duration = round(duration * 2) / 2
    
    # Cap
    if duration > 3.0:
        duration = 3.0
    if duration < 0.5:
        duration = 0.5
    
    return duration


# ============================================
# 2. MAIN TASK GENERATOR FUNCTION
# ============================================

def generate_tasks_from_ml(subjects_data, week_focus='balanced'):
    """
    Generate compact to-do list by randomly selecting 1-2 task types per subject
    
    week_focus options:
    - 'balanced': Equal chance for all task types
    - 'review': Higher chance for review tasks
    - 'exercises': Higher chance for exercises
    - 'past_bac': Higher chance for past BAC
    """
    
    # Task type weights based on week focus
    focus_weights = {
        'balanced': {'review': 0.25, 'exercises': 0.25, 'past_bac': 0.25, 'weakness_focus': 0.25},
        'review': {'review': 0.50, 'exercises': 0.20, 'past_bac': 0.15, 'weakness_focus': 0.15},
        'exercises': {'review': 0.20, 'exercises': 0.50, 'past_bac': 0.15, 'weakness_focus': 0.15},
        'past_bac': {'review': 0.15, 'exercises': 0.15, 'past_bac': 0.55, 'weakness_focus': 0.15}
    }
    
    weights = focus_weights.get(week_focus, focus_weights['balanced'])
    
    tasks = []
    task_id = 0
    MAX_WEEKLY_HOURS = 45
    
    # Past BAC years for randomness
    past_bac_years = ['2024', '2023', '2022', '2021']
    
    for subject_id, data in subjects_data.items():
        coefficient = data.get('coefficient', 2)
        weaknesses = data.get('weaknesses', [])
        review_topics = SUBJECT_TOPICS.get(subject_id, {}).get('review', [])
        
        # Combine all possible topics
        all_topics = list(set(weaknesses + review_topics))
        if len(all_topics) < 2:
            all_topics = all_topics + ['المفاهيم الأساسية', 'تمارين شاملة']
        
        # Shuffle for randomness
        random.shuffle(all_topics)
        
        # Calculate priority
        priority = calculate_priority(data)
        
        # Randomly decide how many tasks for this subject (1 or 2)
        num_tasks = random.choices([1, 2], weights=[0.4, 0.6])[0]
        
        # For high priority subjects (4-5 stars), always give 2 tasks
        if priority >= 4:
            num_tasks = 2
        
        # Select task types based on weights
        selected_types = []
        available_types = list(weights.keys())
        
        for _ in range(num_tasks):
            if not available_types:
                break
            task_type = random.choices(
                available_types, 
                weights=[weights[t] for t in available_types]
            )[0]
            selected_types.append(task_type)
        
        topic_index = 0
        
        for task_type in selected_types:
            # Get topic
            if task_type == 'weakness_focus' and weaknesses and topic_index < len(weaknesses):
                topic = weaknesses[topic_index]
                topic_index += 1
            elif task_type == 'review' and topic_index < len(all_topics):
                topic = all_topics[topic_index]
                topic_index += 1
            elif task_type == 'exercises' and topic_index < len(all_topics):
                topic = all_topics[topic_index]
                topic_index += 1
            elif task_type == 'past_bac':
                topic = random.choice(past_bac_years)
            else:
                topic = 'عام'
            
            # Calculate duration
            duration = calculate_task_duration(data, task_type)
            
            # Apply week focus multiplier to duration
            focus_multipliers = {
                'balanced': 1.0,
                'review': 1.2 if task_type == 'review' else 0.9,
                'exercises': 1.2 if task_type == 'exercises' else 0.9,
                'past_bac': 1.2 if task_type == 'past_bac' else 0.9
            }
            duration = duration * focus_multipliers.get(week_focus, 1.0)
            duration = round(duration * 2) / 2
            
            # Small random variation (±0.25h)
            variation = random.choice([-0.25, 0, 0, 0, 0.25])
            duration = duration + variation
            if duration < 0.5:
                duration = 0.5
            if duration > 4.5:
                duration = 4.5
            duration = round(duration * 2) / 2
            
            # Set icon and display
            if task_type == 'review':
                icon = '📖'
                display_text = f"📖 {topic}"
            elif task_type == 'exercises':
                icon = '✍️'
                display_text = f"✍️ {topic}"
            elif task_type == 'weakness_focus':
                icon = '🎯'
                display_text = f"🎯 {topic}"
            elif task_type == 'past_bac':
                icon = '📝'
                display_text = f"📝 بكالوريا {topic}"
            else:
                icon = '📚'
                display_text = f"📚 {topic}"
            
            # Best time based on task type
            if task_type in ['review', 'weakness_focus', 'past_bac']:
                best_time = 'morning'
            else:
                best_time = 'afternoon'
            
            tasks.append({
                'id': task_id,
                'subject_id': subject_id,
                'subject_name': data['name'],
                'task_type': task_type,
                'icon': icon,
                'description': display_text,
                'duration': duration,
                'priority': priority,
                'splittable': task_type != 'past_bac',
                'best_time': best_time,
                'deadline_day': 'weekend' if task_type == 'past_bac' else 'any',
                'weakness_target': topic if task_type == 'weakness_focus' else None,
                'ml_score': data['ml_score'],
                'is_weak': data.get('weak', False)
            })
            task_id += 1
    
    # Shuffle tasks for variety
    random.shuffle(tasks)
    
    # Sort by priority (highest first) for display
    tasks.sort(key=lambda x: -x['priority'])
    
    # ============================================
    # AUTO-REDUCE if exceeding 45 hours
    # ============================================
    
    total_hours = sum(t['duration'] for t in tasks)
    
    if total_hours > MAX_WEEKLY_HOURS:
        reduction_needed = total_hours - MAX_WEEKLY_HOURS
        
        # Sort tasks by priority (lowest first), then by duration (largest first)
        tasks_to_reduce = [t for t in tasks if t['task_type'] != 'past_bac']
        tasks_to_reduce.sort(key=lambda x: (x['priority'], -x['duration']))
        
        for task in tasks_to_reduce:
            if reduction_needed <= 0:
                break
            
            reduce_by = min(0.5, reduction_needed, task['duration'] - 0.5)
            if reduce_by > 0:
                task['duration'] = round(task['duration'] - reduce_by, 1)
                reduction_needed -= reduce_by
        
        if reduction_needed > 0:
            past_bac_tasks = [t for t in tasks if t['task_type'] == 'past_bac']
            past_bac_tasks.sort(key=lambda x: x['priority'])
            
            for task in past_bac_tasks:
                if reduction_needed <= 0:
                    break
                reduce_by = min(0.5, reduction_needed, task['duration'] - 1.5)
                if reduce_by > 0:
                    task['duration'] = round(task['duration'] - reduce_by, 1)
                    reduction_needed -= reduce_by
    
    return tasks


# ============================================
# 3. WEEK FOCUS SELECTION WIDGET
# ============================================

week_focus_options = ['balanced', 'review', 'exercises', 'past_bac']

week_focus_labels = {
    'balanced': '⚖️ متوازن (جميع أنواع المهام)',
    'review': '📖 تركيز على المراجعة',
    'exercises': '✍️ تركيز على التمارين',
    'past_bac': '📝 تركيز على بكالوريا سابقة'
}

week_focus_dropdown = widgets.Dropdown(
    options=[(week_focus_labels[k], k) for k in week_focus_options],
    value='balanced',
    description='🎯 تركيز الأسبوع:',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='400px', margin='10px 0')
)

# Button to regenerate tasks
regenerate_btn = widgets.Button(
    description='🔄 إعادة توليد المهام (عشوائي)',
    button_style='primary',
    layout=widgets.Layout(width='300px', margin='10px 0')
)

# Output area for tasks
tasks_output = widgets.Output()

# ============================================
# 4. DISPLAY GENERATED TASKS (WITH TASK TYPE COLUMN)
# ============================================

def display_generated_tasks(tasks):
    """Display tasks in a formatted table with task type column"""
    
    if not tasks:
        print("⚠️ لا توجد مهام تم توليدها")
        return
    
    total_hours = sum(t['duration'] for t in tasks)
    total_tasks = len(tasks)
    high_priority = sum(1 for t in tasks if t['priority'] >= 4)
    
    print("=" * 80)
    print("📊 ملخص المهام المُقترحة")
    print("=" * 80)
    print(f"   📋 عدد المهام: {total_tasks}")
    print(f"   ⏱️  إجمالي الساعات: {total_hours:.1f} / 45 ساعة")
    print(f"   🔴 مهام ذات أولوية عالية (4-5): {high_priority}")
    
    if total_hours > 45:
        print(f"   ⚠️ تم التخفيض تلقائياً إلى 45 ساعة")
    
    print("=" * 80)
    
    # Task type names in Arabic
    task_type_names = {
        'review': '📖 مراجعة',
        'exercises': '✍️ تمارين',
        'past_bac': '📝 بكالوريا سابقة',
        'weakness_focus': '🎯 تركيز على نقطة ضعف'
    }
    
    # Create table with task type column
    table_data = []
    for task in tasks:
        stars = '⭐' * task['priority'] + '☆' * (5 - task['priority'])
        
        if task['best_time'] == 'morning':
            time_icon = '🌅'
        else:
            time_icon = '☀️'
        
        # Get task type display
        task_type_display = task_type_names.get(task['task_type'], task['task_type'])
        
        table_data.append({
            '#': task['id'] + 1,
            'المادة': task['subject_name'],
            'نوع المهمة': task_type_display,
            'المهمة': task['description'],
            'المدة': f"{task['duration']}h",
            'الأولوية': stars,
            'الوقت': time_icon,
        })
    
    df = pd.DataFrame(table_data)
    display(HTML(df.to_html(escape=False, index=False)))
    
    # Group by subject
    print("\n" + "=" * 80)
    print("📚 توزيع المهام حسب المادة")
    print("=" * 80)
    
    subject_summary = defaultdict(lambda: {'tasks': 0, 'hours': 0, 'priority': 0})
    for task in tasks:
        subject_summary[task['subject_name']]['tasks'] += 1
        subject_summary[task['subject_name']]['hours'] += task['duration']
        subject_summary[task['subject_name']]['priority'] += task['priority']
    
    for subj_name, data in sorted(subject_summary.items(), key=lambda x: x[1]['priority'], reverse=True):
        bar = '█' * int(data['hours'] * 2)
        print(f"   {subj_name}: {data['tasks']} مهام | {data['hours']:.1f} ساعات {bar}")


# ============================================
# 5. TASK GENERATION AND UPDATE
# ============================================

current_generated_tasks = []

def update_tasks(change=None):
    """Regenerate tasks based on current stream and week focus"""
    global current_generated_tasks
    
    with tasks_output:
        clear_output(wait=True)
        
        # Get current stream subjects
        stream = stream_dropdown.value
        subjects_data = get_subjects_for_stream(stream)
        week_focus = week_focus_dropdown.value
        
        print("🔄 جاري توليد المهام بناءً على...")
        print(f"   🎓 الشعبة: {stream}")
        print(f"   🎯 التركيز: {week_focus_labels.get(week_focus, week_focus)}")
        print(f"   📚 عدد المواد: {len(subjects_data)}")
        print(f"   🎲 توزيع عشوائي (كل مادة تأخذ 1-2 مهام عشوائياً)")
        print()
        
        # Generate tasks
        current_generated_tasks = generate_tasks_from_ml(subjects_data, week_focus)
        
        # Display
        display_generated_tasks(current_generated_tasks)


# Connect the regenerate button
regenerate_btn.on_click(update_tasks)

# Observe stream changes (from Cell 2)
def on_stream_change_for_tasks(change):
    update_tasks()

# Connect to stream dropdown from Cell 2
try:
    stream_dropdown.observe(on_stream_change_for_tasks, names='value')
except NameError:
    print("⚠️ Stream dropdown not found - will use default")


# ============================================
# 6. INITIAL DISPLAY
# ============================================

print("=" * 60)
print("🤖 ML To-Do List Generator (الإصدار النهائي - مدمج)")
print("=" * 60)
print("\n📌 الميزات الجديدة:")
print("   ✅ كل مادة تأخذ 1-2 مهام فقط في الأسبوع")
print("   ✅ أنواع المهام مختارة عشوائياً (مراجعة/تمارين/بكالوريا/تركيز)")
print("   ✅ قائمة مدمجة (15-20 مهمة بدلاً من 50+)")
print("   ✅ الأولوية: 90% معامل + 10% نقطة متوقعة")
print("   ✅ بكالوريا سابقة: وقت ثابت حسب المعامل")
print("   ✅ حد أقصى 45 ساعة/أسبوع (تخفيض تلقائي)")
print("=" * 60)

display(week_focus_dropdown)
display(regenerate_btn)
display(tasks_output)

# Initial generation
update_tasks()


# ============================================
# 7. HELPER FUNCTION TO GET CURRENT TASKS
# ============================================

def get_current_tasks():
    """Return the currently generated tasks (for use in Cell 5)"""
    return current_generated_tasks


print("\n" + "=" * 60)
print("✅ Cell 4 executed successfully!")
print("=" * 60)
print("   📋 قائمة مهام مدمجة (15-20 مهمة)")
print("   📊 تظهر الآن عمود 'نوع المهمة'")
print("   🎲 انقر 'إعادة توليد المهام' للحصول على توزيع مختلف")
print("   📊 اختر تركيز الأسبوع من القائمة أعلاه")
print("=" * 60)
print("\n→ Run Cell 5 next (Interactive To-Do List Editor)")
```

```python
# CELL 5: Interactive To-Do List Editor (CORRECTED)
# ============================================
# Run this after Cell 4
# Allows user to add, delete, edit tasks before generating timetable
# HARD CONSTRAINT: Total hours CANNOT exceed 45 hours
# ============================================

# ============================================
# 1. TODO EDITOR CLASS
# ============================================

class TodoEditor:
    def __init__(self, initial_tasks=None):
        self.tasks = []
        self.widgets = []
        self.output = widgets.Output()
        self.MAX_HOURS = 45
        
        if initial_tasks:
            self.tasks = copy.deepcopy(initial_tasks)
        else:
            self.tasks = []
    
    def get_current_stream_subjects(self):
        """Get list of subjects for the current stream"""
        try:
            stream = stream_dropdown.value
            subjects_data = get_subjects_for_stream(stream)
            return subjects_data
        except:
            return {}
    
    def create_task_row(self, task, index):
        """Create editable widgets for a single task"""
        
        # Checkbox for selection (to delete)
        checkbox = widgets.Checkbox(
            value=False, 
            indent=False,
            layout=widgets.Layout(width='30px')
        )
        
        # Subject dropdown (editable now!)
        subjects_data = self.get_current_stream_subjects()
        subject_options = [(data['name'], subject_id) for subject_id, data in subjects_data.items()]
        
        if not subject_options:
            subject_options = [('📖 اللغة العربية', 'arabic')]
        
        # Find current subject value
        current_subject_value = task['subject_id']
        # Check if current subject exists in options, if not use first
        subject_exists = any(subject_id == current_subject_value for _, subject_id in subject_options)
        if not subject_exists and subject_options:
            current_subject_value = subject_options[0][1]
            task['subject_id'] = current_subject_value
            task['subject_name'] = subject_options[0][0]
        
        subject_dropdown = widgets.Dropdown(
            options=subject_options,
            value=current_subject_value,
            layout=widgets.Layout(width='160px')
        )
        
        # Task type dropdown
        type_options = [
            ('📖 مراجعة', 'review'),
            ('✍️ تمارين', 'exercises'),
            ('📝 بكالوريا سابقة', 'past_bac'),
            ('🎯 تركيز على نقطة ضعف', 'weakness_focus')
        ]
        type_dropdown = widgets.Dropdown(
            options=type_options,
            value=task['task_type'],
            layout=widgets.Layout(width='140px')
        )
        
        # Description (editable text)
        desc_input = widgets.Text(
            value=task['description'],
            placeholder='وصف المهمة',
            layout=widgets.Layout(width='200px')
        )
        
        # Duration slider (0.5 to 4.5 hours, step 0.5)
        duration_slider = widgets.FloatSlider(
            value=task['duration'],
            min=0.5,
            max=4.5,
            step=0.5,
            layout=widgets.Layout(width='120px'),
            readout=False
        )
        duration_label = widgets.Label(value=f"{task['duration']}h", layout=widgets.Layout(width='40px'))
        
        # Priority slider (1-5 stars)
        priority_slider = widgets.IntSlider(
            value=task['priority'],
            min=1,
            max=5,
            layout=widgets.Layout(width='120px'),
            readout=False
        )
        priority_label = widgets.Label(value='⭐' * task['priority'], layout=widgets.Layout(width='70px'))
        
        # Best time dropdown
        best_time_dropdown = widgets.Dropdown(
            options=['morning', 'afternoon', 'evening', 'any'],
            value=task.get('best_time', 'any'),
            layout=widgets.Layout(width='90px')
        )
        
        # Delete button
        delete_btn = widgets.Button(
            description='🗑️',
            button_style='danger',
            layout=widgets.Layout(width='50px')
        )
        
        # Link subject dropdown to task
        def update_subject(change, task_idx=index):
            new_subject_id = change['new']
            if task_idx < len(self.tasks):
                # Find subject name
                subjects_data = self.get_current_stream_subjects()
                if new_subject_id in subjects_data:
                    self.tasks[task_idx]['subject_id'] = new_subject_id
                    self.tasks[task_idx]['subject_name'] = subjects_data[new_subject_id]['name']
                    # Update description to reflect new subject
                    old_desc = self.tasks[task_idx]['description']
                    # Remove old subject name from description if present
                    parts = old_desc.split(' - ')
                    if len(parts) > 1:
                        self.tasks[task_idx]['description'] = f"{parts[0]} - {subjects_data[new_subject_id]['name']}"
        subject_dropdown.observe(update_subject, names='value')
        
        # Link task type dropdown to task
        def update_type(change, task_idx=index):
            new_type = change['new']
            if task_idx < len(self.tasks):
                old_type = self.tasks[task_idx]['task_type']
                self.tasks[task_idx]['task_type'] = new_type
                # Update icon in description
                icon_map = {'review': '📖', 'exercises': '✍️', 'past_bac': '📝', 'weakness_focus': '🎯'}
                new_icon = icon_map.get(new_type, '📚')
                old_desc = self.tasks[task_idx]['description']
                # Update icon at start of description
                if old_desc and len(old_desc) > 0:
                    self.tasks[task_idx]['description'] = new_icon + old_desc[1:]
        type_dropdown.observe(update_type, names='value')
        
        # Link duration slider to label
        def update_duration_label(change, label=duration_label, task_idx=index):
            new_duration = change['new']
            label.value = f"{new_duration:.1f}h"
            if task_idx < len(self.tasks):
                self.tasks[task_idx]['duration'] = new_duration
                self.update_total_label()
        duration_slider.observe(update_duration_label, names='value')
        
        # Link priority slider to label
        def update_priority_label(change, label=priority_label, task_idx=index):
            stars = '⭐' * change['new'] + '☆' * (5 - change['new'])
            label.value = stars
            if task_idx < len(self.tasks):
                self.tasks[task_idx]['priority'] = change['new']
        priority_slider.observe(update_priority_label, names='value')
        
        # Link description to task
        def update_description(change, task_idx=index):
            if task_idx < len(self.tasks):
                self.tasks[task_idx]['description'] = change['new']
        desc_input.observe(update_description, names='value')
        
        # Link best time to task
        def update_best_time(change, task_idx=index):
            if task_idx < len(self.tasks):
                self.tasks[task_idx]['best_time'] = change['new']
        best_time_dropdown.observe(update_best_time, names='value')
        
        # Delete button handler
        def on_delete(b, task_idx=index):
            self.delete_task(task_idx)
        delete_btn.on_click(on_delete)
        
        # Row layout
        row = widgets.HBox([
            checkbox, subject_dropdown, type_dropdown, desc_input, 
            duration_slider, duration_label, priority_slider, priority_label, 
            best_time_dropdown, delete_btn
        ])
        
        return {
            'row': row,
            'checkbox': checkbox,
            'task_index': index,
            'subject_dropdown': subject_dropdown,
            'type_dropdown': type_dropdown,
            'desc_input': desc_input,
            'duration_slider': duration_slider,
            'priority_slider': priority_slider,
            'best_time': best_time_dropdown,
            'delete_btn': delete_btn
        }
    
    def update_total_label(self):
        """Update the total hours display and check constraint"""
        total_hours = sum(t['duration'] for t in self.tasks)
        
        if hasattr(self, 'total_label'):
            self.total_label.value = f"📊 إجمالي الساعات: {total_hours:.1f} / {self.MAX_HOURS} ساعة"
            
            if total_hours > self.MAX_HOURS:
                self.total_label.style = {'color': 'red', 'font-weight': 'bold'}
                self.validate_btn.disabled = True
                self.validate_btn.description = '⚠️ تجاوز الحد الأقصى (45 ساعة) - قلل الساعات'
                if hasattr(self, 'warning_label'):
                    self.warning_label.value = f"⚠️ تحذير: أنت تتجاوز {total_hours - self.MAX_HOURS:.1f} ساعة! قلل المدة أو احذف بعض المهام."
                    self.warning_label.style = {'color': 'red'}
            else:
                self.total_label.style = {'color': 'green', 'font-weight': 'bold'}
                self.validate_btn.disabled = False
                self.validate_btn.description = '✅ تأكيد والانتقال إلى الجدول'
                if hasattr(self, 'warning_label'):
                    self.warning_label.value = f"✅ ضمن الحد المسموح ({total_hours:.1f}/45 ساعة)"
                    self.warning_label.style = {'color': 'green'}
    
    def delete_task(self, index):
        """Delete a task from the list"""
        if 0 <= index < len(self.tasks):
            self.tasks.pop(index)
            self.refresh_display()
    
    def delete_selected_tasks(self):
        """Delete all checked tasks"""
        to_delete = []
        for i, w in enumerate(self.widgets):
            if w['checkbox'].value:
                to_delete.append(i)
        
        # Delete from end to start to avoid index issues
        for i in reversed(to_delete):
            if i < len(self.tasks):
                self.tasks.pop(i)
        
        self.refresh_display()
    
    def add_new_task(self):
        """Add a new empty task - defaults to first subject in current stream"""
        # Get list of available subjects from current stream
        subjects_data = self.get_current_stream_subjects()
        
        if subjects_data:
            default_subject_id = list(subjects_data.keys())[0]
            default_subject_name = subjects_data[default_subject_id]['name']
        else:
            default_subject_id = 'arabic'
            default_subject_name = '📖 اللغة العربية'
        
        new_task = {
            'id': len(self.tasks),
            'subject_id': default_subject_id,
            'subject_name': default_subject_name,
            'task_type': 'review',
            'icon': '📖',
            'description': f"📖 مراجعة - {default_subject_name}",
            'duration': 1.0,
            'priority': 3,
            'splittable': True,
            'best_time': 'morning',
            'deadline_day': 'any',
            'weakness_target': None,
            'ml_score': 12.0,
            'is_weak': False
        }
        
        self.tasks.append(new_task)
        self.refresh_display()
    
    def refresh_display(self):
        """Refresh the entire display"""
        with self.output:
            clear_output(wait=True)
            self.display()
    
    def display(self):
        """Display the full editor"""
        with self.output:
            clear_output(wait=True)
            
            # Header
            header = widgets.HBox([
                widgets.Label('✓', layout=widgets.Layout(width='30px')),
                widgets.Label('المادة', layout=widgets.Layout(width='160px')),
                widgets.Label('النوع', layout=widgets.Layout(width='140px')),
                widgets.Label('الوصف', layout=widgets.Layout(width='200px')),
                widgets.Label('المدة', layout=widgets.Layout(width='120px')),
                widgets.Label('', layout=widgets.Layout(width='40px')),
                widgets.Label('الأولوية', layout=widgets.Layout(width='120px')),
                widgets.Label('', layout=widgets.Layout(width='70px')),
                widgets.Label('الوقت', layout=widgets.Layout(width='90px')),
                widgets.Label('', layout=widgets.Layout(width='50px'))
            ])
            display(header)
            
            # Task rows
            self.widgets = []
            for i, task in enumerate(self.tasks):
                w = self.create_task_row(task, i)
                display(w['row'])
                self.widgets.append(w)
            
            # Total hours label
            total_hours = sum(t['duration'] for t in self.tasks)
            self.total_label = widgets.Label(
                value=f"📊 إجمالي الساعات: {total_hours:.1f} / {self.MAX_HOURS} ساعة",
                layout=widgets.Layout(margin='10px 0', font_weight='bold')
            )
            
            # Warning label
            self.warning_label = widgets.Label(
                layout=widgets.Layout(margin='5px 0')
            )
            
            # Buttons
            add_btn = widgets.Button(description='➕ إضافة مهمة', button_style='success', layout=widgets.Layout(width='150px'))
            delete_selected_btn = widgets.Button(description='🗑️ حذف المحددة', button_style='danger', layout=widgets.Layout(width='150px'))
            self.validate_btn = widgets.Button(description='✅ تأكيد والانتقال إلى الجدول', button_style='primary', layout=widgets.Layout(width='250px'))
            reset_btn = widgets.Button(description='🔄 إعادة تعيين إلى المقترحات', button_style='warning', layout=widgets.Layout(width='200px'))
            
            add_btn.on_click(lambda b: self.add_new_task())
            delete_selected_btn.on_click(lambda b: self.delete_selected_tasks())
            self.validate_btn.on_click(lambda b: self.on_validate())
            reset_btn.on_click(lambda b: self.reset_to_ml_tasks())
            
            buttons = widgets.HBox([add_btn, delete_selected_btn, reset_btn, self.validate_btn])
            display(self.total_label)
            display(self.warning_label)
            display(buttons)
            
            # Update labels with correct values
            self.update_total_label()
            
            # Summary
            if self.tasks:
                avg_priority = sum(t['priority'] for t in self.tasks) / len(self.tasks)
                print(f"\n📊 الملخص: {len(self.tasks)} مهمة | ⭐ متوسط الأولوية: {avg_priority:.1f}")
            else:
                print("📊 لا توجد مهام")
    
    def reset_to_ml_tasks(self):
        """Reset to ML-generated tasks from Cell 4"""
        try:
            self.tasks = copy.deepcopy(current_generated_tasks)
            self.refresh_display()
            print("✅ تم إعادة التعيين إلى المهام المقترحة من ML")
        except:
            print("⚠️ لا توجد مهام ML متاحة")
    
    def on_validate(self):
        """Called when user validates the todo list"""
        total_hours = sum(t['duration'] for t in self.tasks)
        
        if total_hours > self.MAX_HOURS:
            print("\n" + "=" * 60)
            print("❌ لا يمكن المتابعة!")
            print("=" * 60)
            print(f"   إجمالي الساعات ({total_hours:.1f}) يتجاوز الحد الأقصى ({self.MAX_HOURS})")
            print(f"   يرجى تقليل {total_hours - self.MAX_HOURS:.1f} ساعة قبل المتابعة")
            print("=" * 60)
            return
        
        print("\n" + "=" * 60)
        print("✅ تم تأكيد قائمة المهام!")
        print("=" * 60)
        print(f"   عدد المهام: {len(self.tasks)}")
        print(f"   إجمالي الساعات: {total_hours:.1f} / {self.MAX_HOURS}")
        print("=" * 60)
        print("\n→ انتقل إلى Cell 6 لمواصلة بناء الجدول")
    
    def get_validated_tasks(self):
        """Return the current state of tasks"""
        return self.tasks


# ============================================
# 2. CREATE AND DISPLAY EDITOR
# ============================================

print("=" * 60)
print("✏️ Interactive To-Do List Editor")
print("=" * 60)
print("\n📝 تعليمات:")
print("   • يمكنك تعديل المادة، نوع المهمة، الوصف، المدة، الأولوية، الوقت المناسب")
print("   • ✅ اختر المربع لحذف مهمة (ثم اضغط 'حذف المحددة')")
print("   • ➕ أضف مهمة جديدة من زر الإضافة (تظهر تلقائياً لأول مادة في الشعبة)")
print("   • 🔄 يمكنك الرجوع إلى اقتراحات ML في أي وقت")
print("   • ⚠️ الحد الأقصى للساعات: 45 ساعة (لا يمكن تجاوزه)")
print("   • ✅ بعد الانتهاء، اضغط 'تأكيد والانتقال إلى الجدول'")
print("=" * 60)

# Create editor with ML-generated tasks
try:
    editor = TodoEditor(current_generated_tasks)
    display(editor.output)
    editor.display()
except NameError:
    print("⚠️ لم يتم العثور على مهام ML. تأكد من تشغيل Cell 4 أولاً")
    editor = TodoEditor([])
    display(editor.output)
    editor.display()

print("\n" + "=" * 60)
print("✅ Cell 5 executed successfully!")
print("=" * 60)
print("   ✏️ محرر المهام التفاعلي جاهز")
print("   📝 قم بتعديل المهام حسب رغبتك")
print("   ⚠️ لا يمكن تجاوز 45 ساعة")
print("   ✅ بعد الانتهاء، انتقل إلى Cell 6")
print("=" * 60)
```

```python
# CELL 6: Time Slots & Constraints (49 SLOTS - FLEXIBLE VERSION)
# ============================================
# Run this after Cell 5
# 
# NEW DESIGN:
# - 49 total slots (always enough space)
# - Flexible day limits (no hard caps)
# - Lunch break user-selectable (12-13 or 13-14)
# ============================================

# ============================================
# 1. DAYS OF THE WEEK (7 days)
# ============================================

DAYS = ['الأحد', 'الإثنين', 'الثلاثاء', 'الأربعاء', 'الخميس', 'الجمعة', 'السبت']

print("=" * 60)
print("⏰ Time Slots & Constraints (49 Slots - Flexible)")
print("=" * 60)

# ============================================
# 2. LUNCH BREAK SELECTOR (USER CHOICE)
# ============================================

lunch_hour_dropdown = widgets.Dropdown(
    options=[('12:00-13:00', 12), ('13:00-14:00', 13)],
    value=12,
    description='🍽️ وقت الغداء:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px', margin='10px 0')
)

print("📌 سيتم اختيار وقت الغداء لاحقاً في هذه الخلية")

# ============================================
# 3. TIME SLOTS DEFINITION (1-HOUR EACH)
# ============================================

# All possible time slots (08:00 to 18:00) = 10 slots
ALL_TIME_SLOTS = [
    {'name': '08:00-09:00', 'duration': 1.0, 'type': 'morning', 'order': 0},
    {'name': '09:00-10:00', 'duration': 1.0, 'type': 'morning', 'order': 1},
    {'name': '10:00-11:00', 'duration': 1.0, 'type': 'morning', 'order': 2},
    {'name': '11:00-12:00', 'duration': 1.0, 'type': 'morning', 'order': 3},
    {'name': '12:00-13:00', 'duration': 1.0, 'type': 'lunch', 'is_break': True, 'order': 4},
    {'name': '13:00-14:00', 'duration': 1.0, 'type': 'afternoon', 'order': 5},
    {'name': '14:00-15:00', 'duration': 1.0, 'type': 'afternoon', 'order': 6},
    {'name': '15:00-16:00', 'duration': 1.0, 'type': 'afternoon', 'order': 7},
    {'name': '16:00-17:00', 'duration': 1.0, 'type': 'afternoon', 'order': 8},
    {'name': '17:00-18:00', 'duration': 1.0, 'type': 'evening', 'order': 9},
]

# ============================================
# 4. DAY-SPECIFIC SLOT CONFIGURATION
# ============================================

def get_available_slots_for_day(day, lunch_hour=12):
    """Return available time slots for a given day"""
    
    if day == 'الأربعاء':
        # Wednesday: only morning slots (08:00-13:00) = 5 slots
        return [s for s in ALL_TIME_SLOTS if s['order'] < 5]
    
    elif day in ['الجمعة', 'السبت']:
        # Weekend: only morning slots (08:00-12:00) = 4 slots
        return [s for s in ALL_TIME_SLOTS if s['order'] < 4]
    
    else:
        # Weekdays: all slots EXCEPT the lunch hour (which becomes break)
        # If lunch is 12-13, remove that slot; if lunch is 13-14, remove that slot
        slots = []
        for s in ALL_TIME_SLOTS:
            if s.get('is_break', False):
                # This is the 12-13 slot
                if lunch_hour == 12:
                    continue  # Skip 12-13, it's lunch
                else:
                    slots.append(s)  # Keep 12-13 as study slot
            elif s['name'] == f"{lunch_hour:02d}:00-{lunch_hour+1:02d}:00":
                # This is the lunch hour (13-14 if lunch_hour=13)
                continue  # Skip lunch hour
            else:
                slots.append(s)
        return slots

# ============================================
# 5. BUILD SLOT MAPPING
# ============================================

def build_slot_mapping(lunch_hour=12):
    """Build slot mapping based on selected lunch hour"""
    slot_mapping = []
    absolute_idx = 0
    
    for day in DAYS:
        available_slots = get_available_slots_for_day(day, lunch_hour)
        for local_idx, slot in enumerate(available_slots):
            slot_mapping.append({
                'absolute_index': absolute_idx,
                'day': day,
                'day_index': DAYS.index(day),
                'time': slot['name'],
                'duration': slot['duration'],
                'type': slot['type'],
                'order': slot['order']
            })
            absolute_idx += 1
    
    return slot_mapping, absolute_idx

# Initial build with default lunch hour (12)
slot_mapping, TOTAL_SLOTS = build_slot_mapping(12)

print(f"\n📊 إجمالي الفترات المتاحة: {TOTAL_SLOTS} ساعة")
print(f"   (أكثر من الـ 45 ساعة المطلوبة - مساحة كافية لجميع المهام)")

# ============================================
# 6. UPDATE FUNCTION WHEN LUNCH HOUR CHANGES
# ============================================

def update_slots(change=None):
    """Rebuild slot mapping when lunch hour changes"""
    global slot_mapping, TOTAL_SLOTS
    lunch_hour = lunch_hour_dropdown.value
    slot_mapping, TOTAL_SLOTS = build_slot_mapping(lunch_hour)
    
    with slots_output:
        clear_output(wait=True)
        print(f"🍽️ وقت الغداء: {lunch_hour}:00-{lunch_hour+1}:00")
        print(f"📊 إجمالي الفترات المتاحة: {TOTAL_SLOTS} ساعة")
        print("\n📅 الفترات لكل يوم:")
        
        for day in DAYS:
            slots = get_available_slots_for_day(day, lunch_hour)
            slot_names = [s['name'] for s in slots]
            print(f"   {day}: {len(slots)} فترات - {', '.join(slot_names)}")

slots_output = widgets.Output()
lunch_hour_dropdown.observe(update_slots, names='value')

# ============================================
# 7. HARD CONSTRAINTS (MINIMAL - ONLY ESSENTIAL)
# ============================================

HARD_CONSTRAINTS = {
    'lunch_hour': 12,  # Will be updated by dropdown
    'wednesday_afternoon_off': True,
    'weekend_morning_only': True,
    'no_overlap': True,
    'no_cross_day': True,
}

print("\n" + "=" * 60)
print("🔒 HARD CONSTRAINTS (القيود الصارمة)")
print("=" * 60)
print("   • ممنوع تخطي وقت الغداء")
print("   • الأربعاء: دراسة فقط قبل 13:00")
print("   • الجمعة والسبت: دراسة فقط قبل 12:00")
print("   • لا يمكن لمهمتين أن تشغلا نفس الوقت")
print("   • لا يمكن لمهمة أن تمتد عبر منتصف الليل")

# ============================================
# 8. SOFT CONSTRAINTS (FOR FITNESS)
# ============================================

SOFT_WEIGHTS = {
    'daily_balance': 15,      # توزيع متوازن للساعات على الأيام
    'morning_priority': 15,   # المواد عالية الأولوية في الصباح
    'task_splitting': 5,      # تجنب تقسيم المهام
    'consecutive_subject': 3, # تجنب تكرار نفس المادة
}

print("\n" + "=" * 60)
print("⚖️ SOFT CONSTRAINTS (القيود المرنة)")
print("=" * 60)
for key, weight in SOFT_WEIGHTS.items():
    print(f"   • {key}: وزن {weight}")

# ============================================
# 9. DISPLAY
# ============================================

print("\n" + "=" * 60)
print("✅ Cell 6 executed successfully!")
print("=" * 60)
print(f"   📅 أيام الدراسة: {len(DAYS)} أيام")
print(f"   🎯 إجمالي الفترات: {TOTAL_SLOTS} ساعة (≥ 45)")
print(f"   ✅ مساحة كافية لجميع المهام")
print("=" * 60)
print("\n📌 يمكنك تغيير وقت الغداء من القائمة أعلاه")
print("\n→ Run Cell 7 next (Fitness Function)")

# Display lunch selector
display(lunch_hour_dropdown)
display(slots_output)
update_slots()
```

```python
# CELL 7:=run greedy search instead of GA parece que j'en ai marre (HARD DAILY TARGETS)
# ============================================
# Run this after Cell 8
# 
# FIXES:
# 1. Hard daily hour targets (forces even distribution across ALL 7 days)
# 2. Every task scheduled with correct duration (no missing hours)
# 3. No partial scheduling (task only placed if ALL consecutive slots fit)
# ============================================

import math

# ============================================
# 1. GET VALIDATED TASKS FROM EDITOR
# ============================================

print("=" * 60)
print("🎯 Running GA with Validated Tasks")
print("=" * 60)

if 'editor' in globals() and hasattr(editor, 'get_validated_tasks'):
    validated_tasks = editor.get_validated_tasks()
    print(f"📋 Loaded {len(validated_tasks)} tasks from editor")
    
    total_hours = sum(t['duration'] for t in validated_tasks)
    high_priority = sum(1 for t in validated_tasks if t['priority'] >= 4)
    
    print(f"   ⏱️  Total hours needed: {total_hours:.1f}h")
    print(f"   🔴 High priority tasks (4-5): {high_priority}")
else:
    print("⚠️ No tasks found. Creating default tasks...")
    validated_tasks = []
    for i, (sid, data) in enumerate(current_stream_subjects.items()):
        validated_tasks.append({
            'id': i,
            'subject_id': sid,
            'subject_name': data['name'],
            'task_type': 'review',
            'icon': '📖',
            'description': f"📖 مراجعة {data['name']}",
            'duration': 1.5,
            'priority': 4 if data.get('weak', False) else 3,
            'splittable': True,
            'best_time': 'morning',
            'deadline_day': 'any'
        })
    print(f"📋 Created {len(validated_tasks)} default tasks")

print(f"\n✅ Ready to schedule {len(validated_tasks)} tasks")

# ============================================
# 2. DAILY HOUR TARGETS (HARD CONSTRAINTS)
# ============================================

# Target hours per day (will be adjusted based on total hours)
DAILY_TARGETS = {
    'الأحد': 6.5,
    'الإثنين': 6.5,
    'الثلاثاء': 6.5,
    'الأربعاء': 4.0,  # Afternoon off
    'الخميس': 6.5,
    'الجمعة': 3.5,    # Morning only
    'السبت': 3.5,     # Morning only
}

# Normalize targets to match total hours
total_target = sum(DAILY_TARGETS.values())
if total_target != total_hours:
    ratio = total_hours / total_target
    for day in DAILY_TARGETS:
        DAILY_TARGETS[day] = round(DAILY_TARGETS[day] * ratio, 1)

print("\n" + "=" * 60)
print("📊 Daily Hour Targets:")
print("=" * 60)
for day, target in DAILY_TARGETS.items():
    print(f"   {day}: {target}h")

# ============================================
# 3. HELPER FUNCTIONS
# ============================================

def get_task_slots(start_slot, duration):
    """ROUND UP: 1.5h → 2 slots, 0.5h → 1 slot"""
    num_slots = math.ceil(duration)
    return list(range(start_slot, start_slot + num_slots))

def get_day_from_slot(slot_idx):
    """Get day name from slot index"""
    if 0 <= slot_idx < len(slot_mapping):
        return slot_mapping[slot_idx]['day']
    return None

def get_hour_from_slot(slot_idx):
    """Get hour from slot index"""
    if 0 <= slot_idx < len(slot_mapping):
        time = slot_mapping[slot_idx]['time']
        return int(time.split(':')[0])
    return 0

def can_place_task(start_slot, duration, slot_usage, daily_usage):
    """Check if task fits - respects daily targets"""
    num_slots = math.ceil(duration)
    
    if start_slot + num_slots > TOTAL_SLOTS:
        return False
    
    # Track which days will be affected
    days_affected = {}
    
    for i in range(num_slots):
        slot_idx = start_slot + i
        if slot_idx >= len(slot_mapping):
            return False
        
        # Check slot availability
        if slot_usage.get(slot_idx, 0) > 0.01:
            return False
        
        # Check day constraints
        day = get_day_from_slot(slot_idx)
        hour = get_hour_from_slot(slot_idx)
        
        if day == 'الأربعاء' and hour >= 13:
            return False
        if day in ['الجمعة', 'السبت'] and hour >= 12:
            return False
        
        # Track usage per day
        days_affected[day] = days_affected.get(day, 0) + 1
    
    # Check if adding this task would exceed daily targets
    for day, additional_hours in days_affected.items():
        current = daily_usage.get(day, 0)
        target = DAILY_TARGETS.get(day, 8)
        if current + additional_hours > target + 0.5:  # Allow 0.5h flexibility
            return False
    
    return True

def place_task(start_slot, duration, slot_usage, daily_usage, add=True):
    """Place or remove task from slots and daily usage"""
    num_slots = math.ceil(duration)
    sign = 1 if add else -1
    
    for i in range(num_slots):
        slot_idx = start_slot + i
        slot_usage[slot_idx] = slot_usage.get(slot_idx, 0) + sign * 1.0
        
        day = get_day_from_slot(slot_idx)
        if day:
            daily_usage[day] = daily_usage.get(day, 0) + sign * 1.0

def schedule_all_tasks(tasks):
    """Schedule ALL tasks using greedy algorithm with daily targets"""
    num_tasks = len(tasks)
    chromosome = [-1] * num_tasks
    slot_usage = {i: 0.0 for i in range(TOTAL_SLOTS)}
    daily_usage = {day: 0.0 for day in DAYS}
    
    # Sort by priority (highest first), then by duration (largest first)
    sorted_indices = sorted(range(num_tasks), 
                            key=lambda i: (-tasks[i]['priority'], -tasks[i]['duration']))
    
    for task_idx in sorted_indices:
        task = tasks[task_idx]
        duration = task['duration']
        num_slots = math.ceil(duration)
        placed = False
        
        # Try all possible slots
        for start_slot in range(TOTAL_SLOTS - num_slots + 1):
            if can_place_task(start_slot, duration, slot_usage, daily_usage):
                chromosome[task_idx] = start_slot
                place_task(start_slot, duration, slot_usage, daily_usage, add=True)
                placed = True
                break
        
        if not placed:
            print(f"   ⚠️ Could not place: {task['description'][:40]}")
    
    return chromosome, slot_usage, daily_usage

# ============================================
# 4. SCHEDULE ALL TASKS
# ============================================

print("\n" + "=" * 60)
print("📅 Scheduling all tasks with daily targets")
print("=" * 60)

# Try multiple random orders to find best fit
best_chromosome = None
best_daily_usage = None
best_slot_usage = None
best_score = -1

for attempt in range(100):
    # Shuffle task order for variety
    random_tasks = validated_tasks.copy()
    random.shuffle(random_tasks)
    
    chromosome, slot_usage, daily_usage = schedule_all_tasks(random_tasks)
    
    # Check if all tasks scheduled
    if -1 in chromosome:
        continue
    
    # Calculate score (lower variance from targets is better)
    score = 0
    for day, target in DAILY_TARGETS.items():
        actual = daily_usage.get(day, 0)
        diff = abs(actual - target)
        score -= diff * 10  # Penalize deviation from target
    
    if score > best_score:
        best_score = score
        best_chromosome = chromosome
        best_slot_usage = slot_usage
        best_daily_usage = daily_usage

# ============================================
# 5. VERIFY ALL TASKS ARE SCHEDULED
# ============================================

print("\n" + "=" * 60)
print("🔍 Verification")
print("=" * 60)

if best_chromosome is None:
    print("❌ FAILED: Could not schedule all tasks")
    print("   Trying fallback schedule...")
    
    # Fallback: schedule without daily targets
    best_chromosome, best_slot_usage, best_daily_usage = schedule_all_tasks(validated_tasks)

unscheduled = [i for i, g in enumerate(best_chromosome) if g == -1]
if unscheduled:
    print(f"❌ ERROR: {len(unscheduled)} tasks could not be scheduled!")
    for i in unscheduled:
        print(f"   - {validated_tasks[i]['description']}")
else:
    print(f"✅ SUCCESS: All {len(validated_tasks)} tasks are scheduled!")

# Show daily distribution
print("\n📊 Final Daily Distribution:")
for day in DAYS:
    actual = best_daily_usage.get(day, 0)
    target = DAILY_TARGETS.get(day, 0)
    bar = '█' * int(actual * 2)
    status = "✅" if abs(actual - target) <= 1 else "⚠️"
    print(f"   {day}: {actual:.1f}h / {target}h {bar} {status}")

# ============================================
# 6. BUILD TIMETABLE
# ============================================

class ChromosomeWrapper:
    def __init__(self, genes):
        self.genes = genes

best_chromosome_obj = ChromosomeWrapper(best_chromosome)

# Build timetable
timetable = {}
for day in DAYS:
    for slot_info in slot_mapping:
        if slot_info['day'] == day:
            timetable[(day, slot_info['time'])] = None

daily_hours = {day: 0.0 for day in DAYS}
daily_tasks = {day: 0 for day in DAYS}
daily_subjects = {day: [] for day in DAYS}
task_assignments = []

for task_idx, start_slot in enumerate(best_chromosome_obj.genes):
    if start_slot >= 0 and task_idx < len(validated_tasks):
        task = validated_tasks[task_idx]
        duration = task['duration']
        
        task_slots = get_task_slots(start_slot, duration)
        
        for slot_idx in task_slots:
            if slot_idx < len(slot_mapping):
                slot_info = slot_mapping[slot_idx]
                day = slot_info['day']
                time = slot_info['time']
                
                timetable[(day, time)] = task
                daily_hours[day] += 1.0
                daily_tasks[day] += 1
                if task['subject_name'] not in daily_subjects[day]:
                    daily_subjects[day].append(task['subject_name'])
                
                task_assignments.append({
                    'task': task,
                    'day': day,
                    'time': time,
                    'duration': 1.0,
                    'slot_idx': slot_idx
                })

# Store results
final_timetable = timetable
final_task_assignments = task_assignments
final_daily_summary = {
    'hours': daily_hours,
    'tasks': daily_tasks,
    'subjects': daily_subjects
}

print("\n" + "=" * 60)
print("📅 Generated Timetable Summary")
print("=" * 60)

for day in DAYS:
    hours = daily_hours[day]
    tasks_count = daily_tasks[day]
    bar = '█' * int(min(40, hours * 5))
    print(f"   {day}: {hours:.1f}h ({tasks_count} مهام) {bar}")

print(f"\n   📋 Total scheduled: {len(set([a['task']['id'] for a in task_assignments]))}/{len(validated_tasks)} tasks")
print(f"   ⏱️  Total hours scheduled: {sum(daily_hours.values()):.1f}h")

if len(set([a['task']['id'] for a in task_assignments])) == len(validated_tasks):
    print(f"\n   ✅ SUCCESS: All {len(validated_tasks)} tasks are scheduled with correct durations!")
    print(f"   ✅ Tasks distributed across ALL 7 days according to targets!")
else:
    print(f"\n   ❌ FAILURE: {len(validated_tasks) - len(set([a['task']['id'] for a in task_assignments]))} tasks missing")

print("\n" + "=" * 60)
print("✅ Cell 9 executed successfully!")
print("=" * 60)
print("   ✅ EVERY task scheduled with correct duration")
print("   ✅ Tasks distributed across ALL 7 days")
print("   ✅ No missing hours")
print("=" * 60)
print("\n→ Run Cell 10 next (Display Final Timetable)")
```

```python
# CELL 8: Display Final Timetable (CORRECTED - NO DUPLICATE ICONS)
# ============================================
# Run this after Cell 9
# Displays the optimized timetable with all tasks scheduled
# ============================================

print("=" * 60)
print("📅 Your Personalized Study Timetable (7 Days, 45 Hours)")
print("=" * 60)

# ============================================
# 1. HELPER FUNCTIONS FOR CLEAN DISPLAY
# ============================================

def format_task_display(task):
    """Format task for clean display - NO duplicate icons"""
    if not task:
        return "—"
    
    # Get the original description (already has icon at start)
    desc = task['description']
    
    # Priority stars
    stars = '⭐' * task['priority']
    
    # Shorten if too long
    if len(desc) > 30:
        desc = desc[:27] + '...'
    
    # Return as is (description already has the icon)
    return f"{desc}\n{stars}"

def is_consecutive(time1, time2):
    """Check if two times are consecutive"""
    end_hour = int(time1.split('-')[1].split(':')[0])
    start_hour = int(time2.split('-')[0].split(':')[0])
    return end_hour == start_hour

# ============================================
# 2. PREPARE TIMETABLE DATA
# ============================================

# Define time slots to display (08:00 to 18:00)
display_slots = [
    '08:00-09:00', '09:00-10:00', '10:00-11:00', '11:00-12:00',
    '12:00-13:00',  # Lunch
    '13:00-14:00', '14:00-15:00', '15:00-16:00', '16:00-17:00', '17:00-18:00'
]

# Build rows for display
rows = []

for slot in display_slots:
    row = {'الوقت': slot}
    
    for day in DAYS:
        # Wednesday afternoon off
        if day == 'الأربعاء':
            hour = int(slot.split(':')[0])
            if hour >= 13:
                row[day] = '🆓 راحة'
                continue
        
        # Weekend morning only
        if day in ['الجمعة', 'السبت']:
            hour = int(slot.split(':')[0])
            if hour >= 12:
                row[day] = '🆓 راحة'
                continue
        
        # Lunch break
        if slot == '12:00-13:00':
            row[day] = '🍽️ غداء'
            continue
        
        # Get task for this slot
        task = final_timetable.get((day, slot))
        
        if task:
            row[day] = format_task_display(task)
        else:
            row[day] = '—'
    
    rows.append(row)

# Create DataFrame
df = pd.DataFrame(rows)
df = df[['الوقت'] + DAYS]

# Apply styling
def style_timetable(val):
    if val == '🆓 راحة':
        return 'background-color: #e8f5e9; color: #2e7d32; font-weight: bold; text-align: center;'
    elif val == '🍽️ غداء':
        return 'background-color: #fff3e0; color: #e65100; font-weight: bold; text-align: center;'
    elif val and val != '—' and '⭐' in val:
        if '⭐⭐⭐⭐⭐' in val:
            return 'background-color: #ffcdd2; text-align: center;'
        elif '⭐⭐⭐⭐' in val:
            return 'background-color: #ffe0b2; text-align: center;'
        elif '⭐⭐⭐' in val:
            return 'background-color: #c8e6c9; text-align: center;'
        else:
            return 'background-color: #e1bee7; text-align: center;'
    elif val and val != '—':
        return 'background-color: #f3e5f5; text-align: center;'
    return 'background-color: #fafafa; text-align: center;'

print("\n" + "=" * 100)
print("📅 جدولك الدراسي الأسبوعي")
print("=" * 100)

styled_df = df.style.applymap(style_timetable)
display(HTML(styled_df.to_html(escape=False, index=False)))

# ============================================
# 3. DAILY SUMMARY
# ============================================

print("\n" + "=" * 80)
print("📊 الملخص اليومي")
print("=" * 80)

for day in DAYS:
    hours = final_daily_summary['hours'][day]
    tasks_count = final_daily_summary['tasks'][day]
    
    # Visual bar (max 9 hours = 45 chars)
    bar_length = min(45, int(hours * 5))
    bar = '█' * bar_length + '░' * (45 - bar_length)
    
    # Status based on day type
    if day == 'الأربعاء':
        status = "😴 خفيف (بعد الظهر راحة)"
        color = "🟡"
    elif day in ['الجمعة', 'السبت']:
        status = "📚 نهاية الأسبوع (صباحاً فقط)"
        color = "🟢"
    elif hours >= 8:
        status = "⚠️ مكثف - خذ فترات راحة"
        color = "🔴"
    elif hours >= 6:
        status = "✅ جيد"
        color = "🟢"
    elif hours >= 4:
        status = "📚 خفيف"
        color = "🟡"
    else:
        status = "😴 قليل - يمكن زيادة"
        color = "⚪"
    
    print(f"\n📅 {day}:")
    print(f"   {bar} {hours:.1f} ساعات")
    print(f"   📋 {tasks_count} مهام")
    print(f"   {color} {status}")

# ============================================
# 4. TASK DETAILS BY DAY (MERGED CONSECUTIVE SLOTS)
# ============================================

print("\n" + "=" * 80)
print("📋 تفاصيل المهام المجدولة")
print("=" * 80)

# Group tasks by day
tasks_by_day = defaultdict(list)
for assignment in final_task_assignments:
    day = assignment['day']
    tasks_by_day[day].append(assignment)

for day in DAYS:
    if day in tasks_by_day and tasks_by_day[day]:
        print(f"\n📅 {day}:")
        
        # Sort by time
        sorted_tasks = sorted(tasks_by_day[day], key=lambda x: x['time'])
        
        # Group consecutive slots for the same task
        merged_tasks = []
        current = None
        
        for assignment in sorted_tasks:
            task = assignment['task']
            time = assignment['time']
            
            if current is None:
                current = {
                    'task': task,
                    'start_time': time,
                    'end_time': time,
                    'duration': 1.0
                }
            elif current['task']['id'] == task['id'] and is_consecutive(current['end_time'], time):
                current['end_time'] = time
                current['duration'] += 1.0
            else:
                merged_tasks.append(current)
                current = {
                    'task': task,
                    'start_time': time,
                    'end_time': time,
                    'duration': 1.0
                }
        
        if current:
            merged_tasks.append(current)
        
        for i, merged in enumerate(merged_tasks, 1):
            task = merged['task']
            start_hour = merged['start_time'].split('-')[0]
            end_hour = merged['end_time'].split('-')[1]
            time_range = f"{start_hour}-{end_hour}"
            duration = merged['duration']
            stars = '⭐' * task['priority']
            
            # Use description as is (already has icon)
            desc = task['description']
            
            print(f"   {i}. {desc}")
            print(f"      ⏰ {time_range} | ⏱️ {duration:.1f}h | {stars}")

# ============================================
# 5. STUDY TIPS
# ============================================

print("\n" + "=" * 80)
print("💡 نصائح للدراسة")
print("=" * 80)

# Find heavy days
heavy_days = [day for day in DAYS if final_daily_summary['hours'][day] >= 7]
if heavy_days:
    print(f"   ⚠️ أيام مكثفة: {', '.join(heavy_days)} - خذ فترات راحة كل ساعة")

# Find light days
light_days = [day for day in DAYS if final_daily_summary['hours'][day] < 3 and day not in ['الجمعة', 'السبت']]
if light_days:
    print(f"   💡 أيام خفيفة: {', '.join(light_days)} - يمكنك إضافة مراجعة إضافية")

# Wednesday reminder
print(f"   📌 الأربعاء: بعد الظهر راحة إجبارية - استغل الوقت للاسترخاء")

# Weekend reminder
print(f"   📌 الجمعة والسبت: دراسة فقط في الصباح (حتى 12:00)")

# Priority reminder
high_priority_tasks = [a for a in final_task_assignments if a['task']['priority'] >= 4]
if high_priority_tasks:
    unique_high = set([a['task']['subject_name'] for a in high_priority_tasks])
    print(f"   ⭐ مواد عالية الأولوية: {', '.join(list(unique_high)[:5])}")

# Total study time
total_hours = sum(final_daily_summary['hours'].values())
if total_hours >= 40:
    print(f"   ✅ إجمالي ساعات الدراسة: {total_hours:.1f}h - ممتاز، وازن بين الدراسة والراحة")
elif total_hours >= 30:
    print(f"   ✅ إجمالي ساعات الدراسة: {total_hours:.1f}h - جيد، يمكنك زيادة قليلاً")
else:
    print(f"   💡 إجمالي ساعات الدراسة: {total_hours:.1f}h - يمكنك زيادة ساعات الدراسة")

# Study techniques
print(f"\n   🎯 تقنيات دراسة موصى بها:")
print(f"      • تقنية بومودورو: 25 دقيقة دراسة + 5 دقيقة راحة")
print(f"      • ركز على المواد عالية الأولوية في ساعات الصباح")
print(f"      • حل بكالوريات سابقة في عطلة نهاية الأسبوع")

# ============================================
# 6. FINAL SUMMARY
# ============================================

print("\n" + "=" * 80)
print("✅ Cell 10 executed successfully!")
print("=" * 80)
print(f"   📅 تم عرض الجدول الزمني لـ {len(DAYS)} أيام")
print(f"   ⏰ {len(final_task_assignments)} مهمة مجدولة")
print(f"   📊 {sum(final_daily_summary['hours'].values()):.1f} ساعة دراسة")
print("=" * 80)
print("\n🎓 بالتوفيق في البكالوريا!")
print("   اتبع الجدول أعلاه وركز على المواد عالية الأولوية في الصباح")
```
